   - Create comparison tables for side-by-side analysis
   - Use comparisons to ask better follow-up questions

## Section 4: Comparing Columns Using Statistics

### Why Compare Statistics?

Comparing statistics across columns reveals **patterns in your data** that individual analysis misses.

### What to Compare

1. **Average (Mean)** — Which column has highest/lowest typical values?
2. **Spread (Std Dev)** — Which column has most/least variability?
3. **Range (Min to Max)** — Which columns have wider ranges?
4. **Consistency (Mean/Median Difference)** — Which are most/least skewed?

### Example: Employee Survey Comparison

```
satisfaction_score:     Mean=7.2, Median=7.2, Std=1.8
work_life_balance:      Mean=6.8, Median=6.8, Std=2.1
management_support:     Mean=7.5, Median=7.6, Std=1.9

INSIGHTS:
- Management support is highest rated (7.5)
- Work-life balance is lowest rated (6.8)
- Work-life balance has highest variability (2.1)
   → Some people have good balance, others don't
- All three are fairly symmetric (mean ≈ median)
   → No major outlier issues
```

### Build Comparative Tables

Create tables to compare multiple columns at once:

```python
# Calculate statistics for multiple columns
stats = pd.DataFrame({
    'Column': ['satisfaction_score', 'work_life_balance', 'management_support'],
    'Mean': [7.2, 6.8, 7.5],
    'Std': [1.8, 2.1, 1.9],
    'Min': [4, 3, 4],
    'Max': [10, 10, 10]
})
```

This makes it easy to rank and compare columns side-by-side.


````python
# INTERPRETING: Understanding Spread (Standard Deviation)

print("\n" + "=" * 80)
print("INTERPRETATION: Understanding Spread with Standard Deviation")
print("=" * 80)

for col in columns_to_analyze:
    data = df[col]
    mean = data.mean()
    std = data.std()
    data_range = data.max() - data.min()
    cv = (std / mean) * 100  # Coefficient of variation
    
    print(f"\n{col}:")
    print(f"  Mean:       {mean:.2f}")
    print(f"  Std Dev:    {std:.2f}")
    print(f"  Range:      {data.min()} to {data.max()} (width: {data_range})")
    print(f"  Spread %:   {cv:.1f}% of mean")
    
    if std < 1:
        print(f"  ✓ Low variability → Data is consistent, most values are similar")
    elif std < 2:
        print(f"  → Moderate variability → Some spread but not extreme")
    else:
        print(f"  ⚠️  High variability → Large spread, no strong consensus")
````

````python
# INTERPRETING: Comparing Mean vs Median

print("=" * 80)
print("INTERPRETATION: Mean vs Median Analysis")
print("=" * 80)

columns_to_analyze = ['satisfaction_score', 'work_life_balance', 'management_support']

for col in columns_to_analyze:
    mean = df[col].mean()
    median = df[col].median()
    difference = abs(mean - median)
    
    print(f"\n{col}:")
    print(f"  Mean:   {mean:.2f}")
    print(f"  Median: {median:.2f}")
    print(f"  Difference: {difference:.2f}")
    
    if difference < 0.1:
        print(f"  ✓ Nearly equal → Data is symmetric, no major skew")
    elif mean > median:
        print(f"  ⚠️  Mean > Median → Right-skewed, high outliers pull mean up")
    else:
        print(f"  ⚠️  Mean < Median → Left-skewed, low outliers pull mean down")
````

## Section 3: Interpreting Results Correctly

### Critical: Mean vs Median

The **relationship between mean and median** reveals whether your data is skewed (asymmetrical).

#### Case 1: When Mean ≈ Median

```
Case: Symmetric distribution
Mean: 7.2
Median: 7.2
Interpretation: ✓ Data is balanced, no extreme outliers pulling mean
Analysis: Mean is a good representation of typical value
```

#### Case 2: When Mean > Median

```
Case: Right-skewed (pulled up by high outliers)
Mean: 8.5
Median: 6.8
Example: {1, 2, 3, 5, 6, 100}
         Median = 4.5, Mean = 19.5
Interpretation: ⚠️ A few very high values pull the mean up
Analysis: Use MEDIAN as more representative of "typical" value
```

#### Case 3: When Mean < Median

```
Case: Left-skewed (pulled down by low outliers)
Mean: 5.2
Median: 7.8
Interpretation: ⚠️ A few very low values pull the mean down
Analysis: Use MEDIAN as more representative of "typical" value
```

### How to Interpret Standard Deviation

Standard deviation tells you how **consistent** your data is.

#### Small Standard Deviation (Low Variability)

```
Data: {7, 7, 7, 8, 8}
Std Dev: ≈ 0.4
Meaning: Values cluster tightly around the mean (7.4)
Interpretation: ✓ Very consistent data
Analysis: Almost everyone has same satisfaction level
```

#### Large Standard Deviation (High Variability)

```
Data: {1, 3, 7, 9, 10}
Std Dev: ≈ 3.9
Meaning: Values spread widely around the mean
Interpretation: ⚠️ Very inconsistent data
Analysis: No consensus - some very satisfied, some not
```

### Detecting Outliers Without Visualization

Use min/max and compare with quartiles:

```python
Data: {1, 2, 3, 4, 5, 100}
Min: 1, Q1: 2.5, Median: 3.5, Q3: 4.5, Max: 100
Red flag: Max (100) is MUCH larger than Q3 (4.5)
Conclusion: 100 is an outlier!
```

### Common Misinterpretations to Avoid

❌ **Mistake 1:** "Mean is 7, so everyone is satisfied"
✓ **Correct:** Check median and std to see if all values cluster around 7

❌ **Mistake 2:** "Std dev is 2, so it's not important"
✓ **Correct:** For scores 1-10, std of 2 is quite high (20% of range!)

❌ **Mistake 3:** "No missing values means data is clean"
✓ **Correct:** Check min/max for impossible values (negative salaries, future dates)

❌ **Mistake 4:** "Statistics alone tell the whole story"
✓ **Correct:** Statistics guide further investigation, not final conclusions

## Section 1: Understanding Common Summary Statistics

### What Are Summary Statistics?

Summary statistics are **single numbers that describe characteristics of a dataset**.

Instead of looking at all 1000 values, one number tells you the key story about that column.

### The Six Essential Statistics

#### 1️⃣ **Count** — How many values exist?

```
✓ Tells you: Are there missing values?
✓ Formula: Number of non-null values
✓ Example: 45 employees responded (5 missing)
✓ Red flag: If count < total rows, values are missing!
```

#### 2️⃣ **Mean (Average)** — What's the typical value?

```
✓ Tells you: Central tendency of data
✓ Formula: Sum of all values ÷ number of values
✓ Example: Average satisfaction is 7.2 out of 10
✓ Sensitive to: Outliers (very high/low values pull it)
✓ Use for: Understanding typical experience
```

**Mean vs Reality Example:**
- 9 employees satisfied (score: 9)
- 1 employee extremely unhappy (score: 0)
- Mean: (9×9 + 0) ÷ 10 = 8.1
- But 90% are highly satisfied! Mean created wrong impression.

#### 3️⃣ **Median** — What's the middle value?

```
✓ Tells you: Middle point when values are sorted
✓ Formula: Middle value when all values sorted
✓ Example: If scores {1,2,5,7,9} then median is 5
✓ NOT sensitive to: Outliers (ignores extreme values)
✓ Use for: Understanding typical value without outlier distortion
```

**When to prefer median:**
- If 1 person earns $1,000,000 and 99 earn $50,000
- Mean: $59,901 (misleading!)
- Median: $50,000 (realistic!)

#### 4️⃣ **Standard Deviation (Std)** — How spread out is the data?

```
✓ Tells you: How far values typically vary from the mean
✓ Small std: Values cluster closely around the mean (consistent)
✓ Large std: Values spread widely (high variability)
✓ Example: 
   Scenario A: scores {8,8,8,9,9} → std ≈ 0.5 (tight cluster)
   Scenario B: scores {1,3,8,9,10} → std ≈ 3.9 (spread out)
✓ Use for: Understanding data consistency
```

#### 5️⃣ **Minimum (Min)** — What's the smallest value?

```
✓ Tells you: Lowest value in the column
✓ Use for: Checking for unrealistic values
✓ Red flags:
   - Negative salaries?
   - Impossible dates?
   - Out-of-range scores?
```

#### 6️⃣ **Maximum (Max)** — What's the largest value?

```
✓ Tells you: Highest value in the column
✓ Use for: Checking for outliers or unrealistic values
✓ Red flags:
   - Salary too high?
   - Score exceeds expected range?
   - Future dates?
```

### How Statistics Describe Data

**Example: Satisfaction Score Data**

```
Raw values: {4, 5, 6, 7, 7, 8, 8, 8, 9, 10}

✓ Count: 10 (no missing values)
✓ Mean: 7.2 (average satisfaction)
✓ Median: 7.5 (middle value)
✓ Std: 1.75 (moderate spread, fairly consistent)
✓ Min: 4 (one person quite dissatisfied)
✓ Max: 10 (one person very satisfied)

INTERPRETATION: Most people are satisfied (7-8 range) with some variation.
```

### Quick Decision Guide

| Question | Statistic to Use |
|----------|------------------|
| "How many values?" | Count |
| "What's typical?" | Mean OR Median |
| "Is data consistent?" | Standard Deviation |
| "What's the range?" | Min and Max |
| "Are there outliers?" | Compare Mean vs Median, look at Min/Max |
