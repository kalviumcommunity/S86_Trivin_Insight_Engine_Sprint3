## 1. Using For Loops for Iteration

For loops are ideal when you know the sequence you're iterating over. They automatically handle iteration and termination.

### 1.1 Iterating Over a Range

The `range()` function generates a sequence of numbers. This is perfect for counting or repeating actions a specific number of times.

```python
# Example: Process data points 0 to 4
print("Processing data points:")
for i in range(5):
    print(f"Data point {i}: Value = {i * 10}")
```

## 2. Using While Loops for Condition-Based Repetition

While loops continue as long as a condition is True. They're useful when you don't know in advance how many iterations are needed.

### 2.1 Basic While Loop with Counter

```python
# Example: Process batches while count < 5
count = 0

print("Processing batches:")
while count < 5:
    print(f"  Processing batch {count}")
    count += 1  # CRITICAL: Always update the condition variable!

print(f"Completed {count} batches")
```

### 2.2 While Loop with Accumulation

```python
# Example: Accumulate until reaching a target
total = 0
iteration = 1
target = 100

print(f"Accumulating until total reaches {target}:")
while total < target:
    total += iteration * 5
    print(f"  Iteration {iteration}: Total = {total}")
    iteration += 1

print(f"Reached target after {iteration - 1} iterations")
```

### 2.3 Processing Data Until Condition is Met

```python
# Example: Process queue items until threshold
data_queue = [12, 34, 56, 78, 90, 23, 45, 67, 89, 10]
processed = 0
index = 0
threshold = 200

print(f"Processing until sum reaches {threshold}:")
while processed < threshold and index < len(data_queue):
    processed += data_queue[index]
    print(f"  Item {index + 1}: {data_queue[index]} (Total: {processed})")
    index += 1

print(f"Stopped after processing {index} items")
```

## 3. Controlling Loop Flow with Break and Continue

Control statements allow you to manage loop execution more precisely.

### 3.1 Using Break to Exit Early

```python
# Example: Find first error in data
data_records = [
    {"id": 1, "status": "valid"},
    {"id": 2, "status": "valid"},
    {"id": 3, "status": "error"},
    {"id": 4, "status": "valid"},
    {"id": 5, "status": "error"}
]

print("Searching for first error:")
for record in data_records:
    print(f"  Checking record {record['id']}: {record['status']}")
    if record['status'] == "error":
        print(f"  ! First error found at record {record['id']}")
        break  # Stop once we find the first error
```

### 3.2 Using Continue to Skip Iterations

```python
# Example: Process only positive values
values = [10, -5, 20, -3, 30, 0, -8, 40]

print("Processing only positive values:")
for value in values:
    if value <= 0:
        print(f"  Skipping non-positive: {value}")
        continue  # Skip to next iteration
    
    print(f"  Processing positive value: {value}")
```

### 3.3 Combining Break and Continue

```python
# Example: Process valid scores until reaching target
scores = [45, -10, 67, 89, 101, 78, 92, 88]
valid_scores = []
target_score = 90

print("Processing valid scores:")
for score in scores:
    # Skip invalid scores (out of range 0-100)
    if score < 0 or score > 100:
        print(f"  Skipping invalid: {score}")
        continue
    
    # Process valid score
    valid_scores.append(score)
    print(f"  Valid score: {score}")
    
    # Stop if we reach target
    if score >= target_score:
        print(f"  ! Target reached: {score}")
        break

print(f"Processed {len(valid_scores)} valid scores: {valid_scores}")
```

## 4. Avoiding Infinite Loops

Infinite loops occur when the loop condition never becomes False. This can freeze your program!

### 4.1 Common Mistake: Forgetting to Update Loop Variable

```python
# INCORRECT - Would cause infinite loop (DO NOT RUN):
# count = 0
# while count < 5:
#     print(f"Count: {count}")
#     # Missing: count += 1  <- This causes infinite loop!

# CORRECT - Always update the loop variable:
count = 0
while count < 5:
    print(f"Count: {count}")
    count += 1  # This makes the condition eventually False
```

### 4.2 Using Safety Counters

```python
# Example: Using a safety limit to prevent infinite loops
max_iterations = 10
iteration = 0

print(f"Processing with safety limit of {max_iterations}:")
while True:  # Potentially infinite
    iteration += 1
    print(f"  Iteration {iteration}")
    
    # Safety check
    if iteration >= max_iterations:
        print("  ! Safety limit reached")
        break
    
    # Simulating a condition that might never occur
    if iteration == 7:
        print(f"  Condition met at iteration {iteration}")
        break
```

### 4.3 Ensuring Loop Conditions Can Change

```python
# Example: Processing a queue that gets smaller
queue = [10, 20, 30, 40, 50]

print("Processing queue:")
while len(queue) > 0:  # Condition: queue is not empty
    item = queue.pop(0)  # Remove item - changes the condition!
    print(f"  Processed: {item}, Remaining: {len(queue)}")

print("Queue is empty - loop terminated correctly")
```

### 4.4 Infinite Loop Prevention Checklist

✓ Always update loop variables inside the loop  
✓ Ensure loop conditions can eventually become False  
✓ Use for loops when possible (they terminate automatically)  
✓ Add safety counters for complex while loops  
✓ Test loops with small examples first  
✓ Use break statements as emergency exits  
✓ Keep loop logic simple and readable  

## 5. Practical Data Workflow Examples

Let's apply loops to real-world data processing scenarios.

### 5.1 Data Cleaning

```python
# Example: Clean data by removing nulls and error codes
raw_data = [100, 200, -999, 150, None, 180, -999, 220]
cleaned_data = []

print("Cleaning data:")
for value in raw_data:
    if value is None or value == -999:
        print(f"  Skipping invalid: {value}")
        continue
    cleaned_data.append(value)
    print(f"  Kept valid: {value}")

print(f"\nOriginal size: {len(raw_data)}")
print(f"Cleaned size: {len(cleaned_data)}")
print(f"Cleaned data: {cleaned_data}")
```

### 5.2 Data Transformation

```python
# Example: Convert temperatures from Celsius to Fahrenheit
celsius_temps = [0, 10, 20, 25, 30, 35]
fahrenheit_temps = []

print("Converting Celsius to Fahrenheit:")
for celsius in celsius_temps:
    fahrenheit = (celsius * 9/5) + 32
    fahrenheit_temps.append(fahrenheit)
    print(f"  {celsius}°C = {fahrenheit}°F")

print(f"\nOriginal (C): {celsius_temps}")
print(f"Converted (F): {fahrenheit_temps}")
```

### 5.3 Data Aggregation

```python
# Example: Aggregate sales by region
sales_data = [
    {"region": "North", "amount": 1000},
    {"region": "South", "amount": 1500},
    {"region": "North", "amount": 1200},
    {"region": "East", "amount": 800},
    {"region": "South", "amount": 900},
    {"region": "North", "amount": 1100}
]

regional_totals = {}

for sale in sales_data:
    region = sale["region"]
    amount = sale["amount"]
    
    if region not in regional_totals:
        regional_totals[region] = 0
    
    regional_totals[region] += amount

print("Regional Sales Totals:")
for region, total in regional_totals.items():
    print(f"  {region}: ${total}")
```

### 5.4 Batch Processing

```python
# Example: Process records in batches
records = list(range(1, 16))  # 15 records
batch_size = 5
batch_number = 1

print(f"Processing {len(records)} records in batches of {batch_size}:")

index = 0
while index < len(records):
    batch = records[index:index + batch_size]
    print(f"  Batch {batch_number}: {batch}")
    batch_number += 1
    index += batch_size

print("All batches processed")
```

## Summary and Key Takeaways

### What You Learned:

✓ **For Loops:** Iterate over known sequences (lists, ranges, dictionaries)  
✓ **While Loops:** Repeat based on conditions that may change during execution  
✓ **Break:** Exit loops early when a condition is met  
✓ **Continue:** Skip specific iterations while continuing the loop  
✓ **Infinite Loops:** Always update loop variables and use safety mechanisms  
✓ **Data Workflows:** Apply loops to cleaning, transforming, and aggregating data  

### When to Use Which Loop:

| Scenario | Use |
|----------|-----|
| Processing a list or collection | `for loop` |
| Repeating N times | `for loop with range()` |
| Condition-based repetition | `while loop` |
| Unknown number of iterations | `while loop` |
| Iterating with index | `for loop with enumerate()` |
