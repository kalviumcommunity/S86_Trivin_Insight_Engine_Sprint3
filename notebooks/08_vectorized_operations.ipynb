## Section 1: Understanding Loop-Based vs Vectorized Code

### The Difference

Loop-based code explicitly iterates over each element, telling Python *how* to perform operations.

Vectorized code tells NumPy to apply operations to entire arrays at once, letting the library handle the iteration internally in optimized C code.

### Example: Squaring Each Element of an Array

Let's compare the two approaches:


````python
import numpy as np

# Create a simple array
numbers = np.array([1, 2, 3, 4, 5])
print("Original array:", numbers)
````

````python
# APPROACH 1: Loop-Based (Explicit iteration)
squared_loop = []
for num in numbers:
    squared_loop.append(num ** 2)

squared_loop = np.array(squared_loop)
print("\nLoop-based result:", squared_loop)
print("Code: 4 lines for simple operation")
````

````python
# APPROACH 2: Vectorized (Let NumPy handle it)
squared_vectorized = numbers ** 2
print("Vectorized result: ", squared_vectorized)
print("Code: 1 line for same operation")
print("\nResults are identical:", np.array_equal(squared_loop, squared_vectorized))
````

### Comparison Summary

| Aspect | Loop-Based | Vectorized |
|--------|-----------|-----------|
| **Lines of Code** | 4 lines | 1 line |
| **Readability** | Verbose | Concise |
| **Performance** | Slower (Python loops) | Faster (NumPy's C code) |
| **Intent** | *How* to do it | *What* to accomplish |
| **Scalability** | Degrades with size | Scales efficiently |

**Key Insight:** Vectorization tells NumPy your intent, and it handles the efficient execution. The loop is still there, but it runs in optimized C code under the hood!


## Section 2: Applying Vectorized Arithmetic Operations

### Basic Arithmetic Operations

NumPy allows you to apply arithmetic operations (+, -, *, /, **) to entire arrays at once. Each operation is applied element-wise.

### Example 1: Basic Arithmetic Operations


````python
# Create sample arrays
prices = np.array([10.0, 20.0, 30.0, 40.0])
quantities = np.array([2, 3, 1, 5])

print("Prices:   ", prices)
print("Quantities:", quantities)

# ADDITION: Add 5 to each price
prices_increased = prices + 5
print("\nAdd 5 to each price:", prices_increased)

# SUBTRACTION: Reduce prices by 2
prices_reduced = prices - 2
print("Reduce by 2:", prices_reduced)

# MULTIPLICATION: Calculate total cost (element-wise)
total_costs = prices * quantities
print("\nElement-wise multiplication (price × quantity):")
print("Total costs:", total_costs)

# DIVISION: Calculate price per unit (if we divide by 2)
half_prices = prices / 2
print("\nDivide each price by 2:", half_prices)

# EXPONENTIATION: Square the quantities
quantities_squared = quantities ** 2
print("\nSquare each quantity:", quantities_squared)
````

### Example 2: Complex Vectorized Expressions

You can combine multiple operations in a single expression to create more complex calculations:


````python
# Scenario: Apply a 10% discount and add 5% tax
original_prices = np.array([100.0, 150.0, 200.0, 75.0])
print("Original prices:", original_prices)

# Complex expression: (price - 10% discount) + 5% tax
# Step 1: Apply 10% discount → multiply by 0.9
# Step 2: Add 5% tax → multiply the result by 1.05
final_prices = (original_prices * 0.9) * 1.05

print("After 10% discount + 5% tax:", final_prices)

# Even more complex: Convert Celsius to Fahrenheit for multiple temperatures
celsius = np.array([0, 10, 20, 30, 100])
fahrenheit = (celsius * 9/5) + 32

print("\nCelsius to Fahrenheit conversion:")
print("Celsius:   ", celsius)
print("Fahrenheit:", fahrenheit)

# Calculate area of circles with different radii
radii = np.array([1.0, 2.0, 3.0, 5.0])
areas = np.pi * radii ** 2

print("\nAreas of circles:")
print("Radii:", radii)
print("Areas:", areas)
````

### Example 3: Broadcasting (Operating with Different Shapes)

NumPy's broadcasting allows operations between arrays of different shapes. A smaller array is "broadcast" across the larger one:


````python
# Example: Subtract a scalar from each element
test_scores = np.array([85, 90, 78, 92, 88])
curve_adjustment = 5

# The scalar is broadcast to match the array shape
adjusted_scores = test_scores + curve_adjustment
print("Original scores:  ", test_scores)
print("Add 5 points:     ", adjusted_scores)

# Example: Multiply each element of an array by a scalar
data = np.array([1, 2, 3, 4, 5])
multiplier = 10

result = data * multiplier
print("\nOriginal data:    ", data)
print("Multiply by 10:   ", result)

# Example: Multiply along a column
# Imagine you have prices for different products and quantities to buy
products = np.array([10.0, 20.0, 30.0])  # prices of 3 products
quantities_per_visit = np.array([2, 3, 1])  # quantities for this purchase

total = products * quantities_per_visit
print("\nProduct prices:   ", products)
print("Quantities:       ", quantities_per_visit)
print("Total per product:", total)
````

## Section 3: Using Vectorized Comparisons and Conditions

### Comparison Operators Without Loops

Instead of looping to check each element, use comparison operators directly on arrays. These produce boolean arrays that indicate True or False for each element.

### Comparison Operators

| Operator | Meaning | Example |
|----------|---------|---------|
| `==` | Equal to | `array == 5` |
| `!=` | Not equal to | `array != 0` |
| `<` | Less than | `array < 10` |
| `>` | Greater than | `array > 20` |
| `<=` | Less than or equal | `array <= 100` |
| `>=` | Greater than or equal | `array >= 50` |

### Example 1: Simple Comparisons


````python
grades = np.array([75, 85, 92, 68, 88, 78, 95, 72])
passing_grade = 80

print("Grades:", grades)
print("Passing grade:", passing_grade)

# Check which grades are >= passing grade
passing = grades >= passing_grade
print("\nGrades >= 80:", passing)
print("Boolean array type:", type(passing))

# Check which grades are below passing
failing = grades < passing_grade
print("\nGrades < 80:", failing)

# Check which grades equal 80
exactly_80 = grades == 80
print("\nGrades == 80:", exactly_80)

# Check which grades are NOT 80
not_80 = grades != 80
print("\nGrades != 80:", not_80)
````

### Example 2: Filtering Arrays Using Boolean Conditions

Boolean arrays can be used to filter data. The boolean values tell NumPy which elements to keep:


````python
# Extract only passing grades
passing_grades = grades[grades >= 80]
print("Passing grades (>= 80):", passing_grades)

# Extract only failing grades
failing_grades = grades[grades < 80]
print("Failing grades (< 80): ", failing_grades)

# Count how many passed and failed
num_passed = np.sum(grades >= 80)  # True = 1, so sum counts them
num_failed = np.sum(grades < 80)
print(f"\nNumber passed: {num_passed}")
print(f"Number failed: {num_failed}")
print(f"Pass rate: {num_passed / len(grades) * 100:.1f}%")
````

### Example 3: Multiple Conditions Using Logical Operators

Use `&` (and), `|` (or), and `~` (not) to combine conditions:


````python
temperatures = np.array([5, 15, 25, 32, 18, 28, 10, 35, 22])
print("Temperatures:", temperatures)

# Find temperatures between 20 and 30 (inclusive)
# Use & for AND
moderate = (temperatures >= 20) & (temperatures <= 30)
print("\nTemperatures between 20-30:", temperatures[moderate])

# Find temperatures < 10 or > 30 (extreme temperatures)
# Use | for OR
extreme = (temperatures < 10) | (temperatures > 30)
print("Extreme temperatures (< 10 or > 30):", temperatures[extreme])

# Find temperatures NOT between 15 and 25
# Use ~ for NOT
not_mild = ~((temperatures >= 15) & (temperatures <= 25))
print("NOT between 15-25:", temperatures[not_mild])
````

## Section 4: Avoiding Common Vectorization Mistakes

### Mistake 1: Mixing Incompatible Array Shapes

When operations involve arrays of different shapes, NumPy's broadcasting rules must be compatible. Not all combinations work!


````python
arr1 = np.array([1, 2, 3])
arr2 = np.array([10, 20, 30, 40])

print("Array 1 shape:", arr1.shape)
print("Array 2 shape:", arr2.shape)

# MISTAKE: These arrays can't be broadcast together
try:
    result = arr1 + arr2  # This will cause an error
except ValueError as e:
    print("\n❌ ERROR:", e)
    print("   Arrays with shapes (3,) and (4,) cannot be added")

print("\n✓ SOLUTION: Make arrays compatible in shape")
arr2_trimmed = arr2[:3]  # Take first 3 elements
print("Array 2 (trimmed) shape:", arr2_trimmed.shape)
result = arr1 + arr2_trimmed
print("Successfully added:", result)
````

### Mistake 2: Using `and`/`or` Instead of `&`/`|` for Array Conditions

Python's `and` and `or` keywords don't work with arrays. Use `&` and `|` instead.


````python
values = np.array([15, 25, 5, 35, 20])
print("Values:", values)

# MISTAKE: Using 'and' with arrays
print("\n❌ MISTAKE: Using 'and'")
try:
    # This looks right, but won't work as expected
    result = values[(values > 10) and (values < 30)]
except Exception as e:
    print(f"   ERROR: {type(e).__name__}: {e}")

# CORRECT: Using '&' (bitwise AND)
print("\n✓ CORRECT: Using '&' for arrays")
result = values[(values > 10) & (values < 30)]
print("Values between 10 and 30:", result)

# Another example: Find values < 10 OR > 30
print("\n✓ Using '|' (OR) for arrays")
result = values[(values < 10) | (values > 30)]
print("Values < 10 or > 30:", result)
````

### Mistake 3: Premature Optimization - Clarity Over Speed

Vectorization is great for performance, but **clarity and correctness come first**. Don't sacrifice readability for a small speed gain.


````python
# Example: Calculate moving average
data = np.array([10, 15, 20, 18, 22, 25, 23, 28, 26, 30])

# APPROACH 1: Clear and readable (slightly slower)
print("APPROACH 1: Clear and readable")
moving_avg_clear = []
for i in range(1, len(data)):
    avg = (data[i-1] + data[i]) / 2
    moving_avg_clear.append(avg)
moving_avg_clear = np.array(moving_avg_clear)
print("Result:", moving_avg_clear)

# APPROACH 2: Vectorized but less obvious
print("\nAPPROACH 2: Vectorized but less clear")
moving_avg_fast = (data[:-1] + data[1:]) / 2
print("Result:", moving_avg_fast)

# Both produce the same result
print("\nResults are identical:", np.allclose(moving_avg_clear, moving_avg_fast))

print("\n💡 LESSON: Use Approach 1 if the logic is complex.")
print("   Use Approach 2 only if performance matters AND you've profiled it.")
print("   Always prefer clarity first, optimization second.")
````

### Mistake 4: Not Understanding Result Types

Different operations produce different types. Watch out for unexpected results!


````python
integers = np.array([10, 20, 30, 40])
print("Original array (integers):", integers, "| dtype:", integers.dtype)

# Division produces floats
divided = integers / 2
print("After division:          ", divided, "| dtype:", divided.dtype)

# Comparison produces booleans
comparison = integers > 25
print("After comparison (>25):  ", comparison, "| dtype:", comparison.dtype)

# Common issue: Integer division
print("\n❌ MISTAKE: Integer division with // produces integers")
int_division = integers // 3
print("Using //:               ", int_division, "| dtype:", int_division.dtype)

print("\n✓ CORRECT: Regular division / produces floats")
float_division = integers / 3
print("Using /:                ", float_division, "| dtype:", float_division.dtype)

# Be aware of type conversión
print("\n💡 Tip: Use dtype to check array type")
print("   Check dtype before operations requiring specific types")
````

## Key Takeaways: Vectorization Best Practices

### ✓ DO:

1. **Use array operations instead of loops** when working with NumPy arrays
   - `array * 2` instead of `[x * 2 for x in array]`
   
2. **Use comparison operators on arrays** to create boolean masks
   - `array[array > 5]` to filter

3. **Use `&` and `|` for logical conditions** on arrays (not `and`/`or`)
   - `(arr > 10) & (arr < 20)` for multiple conditions

4. **Prefer clarity over micro-optimizations**
   - If the vectorized version is hard to read, explain it with comments

5. **Check array shapes and dtypes** before operations
   - Use `.shape` and `.dtype` to understand your data

6. **Let NumPy handle the iteration** - that's its job!

### ✗ DON'T:

1. **Don't mix incompatible array shapes** without understanding broadcasting
2. **Don't use `and`/`or` with arrays** - use `&`/`|` instead
3. **Don't optimize prematurely** - focus on correctness first
4. **Don't ignore data types** - division and comparison matter
5. **Don't write loops when vectorization is straightforward**
6. **Don't force vectorization** if a loop is clearer

### Remember:

**Vectorization is the NumPy way of thinking:**
- Tell NumPy *what* you want, not *how* to do it
- Operations happen element-wise automatically
- Code is faster, cleaner, and more Pythonic
