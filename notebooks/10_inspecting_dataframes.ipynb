## Why Inspection Matters

**Inspection is the most important step after loading data.**

Common mistakes avoided by proper inspection:
- Starting analysis without understanding the data
- Misinterpreting column data types
- Missing hidden null values
- Drawing conclusions from incomplete information

**Think of inspection as reading the dataset before using it.**

## 1. Inspecting Data with head()

### Purpose: Quick Visual Preview

The `head()` method shows the **first few rows** of your DataFrame.

**What you can learn from head():**
- Column names
- Sample values for each column
- Data alignment and formatting
- General structure at a glance

**Default behavior:** Shows the first 5 rows

````python
# Display the first 5 rows (default)
print("=" * 80)
print("PREVIEW: First 5 rows using head()")
print("=" * 80)

df.head()
````

````python
# You can specify a different number of rows
print("=" * 80)
print("PREVIEW: First 10 rows using head(10)")
print("=" * 80)

df.head(10)
````

````python
# Also useful: tail() shows the LAST few rows
print("=" * 80)
print("PREVIEW: Last 5 rows using tail()")
print("=" * 80)

df.tail()
````

### What Did We Learn from head()?

From the preview above, we can observe:
- **Column names**: employee_id, department, survey_date, satisfaction_score, etc.
- **Data types (visual guess)**: 
  - employee_id appears to be numeric
  - department appears to be text/categorical
  - satisfaction_score appears to be numeric
  - response_text contains free-form text
- **Sample values**: We can see actual employee survey responses
- **No obvious formatting issues** in the first few rows

**Key Point:** head() gives you a quick visual check, but it doesn't tell you about data types, missing values, or overall structure. That's where info() comes in!

---

## 2. Inspecting Structure with info()

### Purpose: Understand DataFrame Structure and Health

The `info()` method provides a **comprehensive structural summary**.

**What you can learn from info():**
- Total number of rows and columns
- Column names
- **Data types** for each column (int64, float64, object, etc.)
- **Non-null counts** (reveals missing values)
- Memory usage

**This is the most important inspection method for understanding data structure.**

````python
# Get detailed structural information
print("=" * 80)
print("STRUCTURE: DataFrame Info")
print("=" * 80)

df.info()
````

### Understanding the info() Output

Let's break down what info() tells us:

1. **RangeIndex**: The index (row labels) - shows how many rows exist
2. **Data columns**: Total number of columns
3. **Column details**:
   - **Column name**
   - **Non-Null Count**: How many values are NOT missing
   - **Dtype**: Data type (int64, float64, object, datetime64, etc.)

4. **dtypes summary**: Quick count of each data type
5. **Memory usage**: How much RAM the DataFrame uses

### Critical Insights:

**If Non-Null Count < Total Rows** → Missing values exist in that column!

**Data type matters:**
- `int64` → Integer numbers (1, 2, 3)
- `float64` → Decimal numbers (1.5, 2.7)
- `object` → Text/strings or mixed types
- `datetime64` → Dates and times

**Common issues detected by info():**
- Numeric columns stored as 'object' (needs conversion)
- Date columns not recognized as datetime
- Missing values in critical columns

````python
# Let's check for missing values explicitly
print("=" * 80)
print("MISSING VALUES: Count per Column")
print("=" * 80)

missing_counts = df.isnull().sum()
print(missing_counts)

print("\n" + "=" * 80)
print("MISSING VALUES: Percentage per Column")
print("=" * 80)

missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent.round(2))
````

## 3. Summarizing Data with describe()

### Purpose: Statistical Summary of Numeric Columns

The `describe()` method provides **statistical summaries** for numeric data.

**What you can learn from describe():**
- Count of non-null values
- Mean (average)
- Standard deviation (spread)
- Minimum value
- 25th percentile (Q1)
- 50th percentile (median)
- 75th percentile (Q3)
- Maximum value

**Default behavior:** Only shows numeric columns

````python
# Get statistical summary of numeric columns
print("=" * 80)
print("STATISTICS: Numeric Column Summary")
print("=" * 80)

df.describe()
````

### Understanding the describe() Output

**What each row means:**

1. **count**: Number of non-missing values
   - If count < total rows → missing values exist

2. **mean**: Average value
   - Sensitive to outliers

3. **std**: Standard deviation (how spread out the data is)
   - High std = high variability
   - Low std = values close to the mean

4. **min**: Smallest value
   - Check for unrealistic values (e.g., negative ages)

5. **25% (Q1)**: First quartile - 25% of values are below this

6. **50% (median)**: Middle value when data is sorted
   - Not affected by outliers like mean is

7. **75% (Q3)**: Third quartile - 75% of values are below this

8. **max**: Largest value
   - Check for unrealistic values or outliers

### What to Look For:

**Potential outliers:**
- If max is much larger than 75% → possible high outlier
- If min is much smaller than 25% → possible low outlier

**Data range:**
- Check if min/max make sense for the variable
- Example: Satisfaction scores should be within expected range (e.g., 1-10)

**Distribution shape:**
- If mean ≈ median → roughly symmetric distribution
- If mean > median → right-skewed (pulled by high values)
- If mean < median → left-skewed (pulled by low values)

````python
# You can also describe categorical columns using include='object'
print("=" * 80)
print("STATISTICS: Categorical Column Summary")
print("=" * 80)

df.describe(include='object')
````

### Understanding describe() for Categorical Data

For object/categorical columns, describe() shows:

1. **count**: Number of non-missing values
2. **unique**: Number of unique/distinct values
3. **top**: Most frequently occurring value
4. **freq**: How many times the top value appears

**This is useful for:**
- Understanding category distributions
- Identifying highly imbalanced categories
- Detecting unexpected values

````python
# Get summary for ALL columns (numeric and categorical)
print("=" * 80)
print("STATISTICS: All Columns Summary")
print("=" * 80)

df.describe(include='all')
````

## 4. Knowing When to Use Each Method

### Building Inspection Intuition

Each inspection method answers a different question:

| Method | Question It Answers | When to Use |
|--------|-------------------|-------------|
| **head()** | "What does the data look like?" | Quick visual preview, check formatting |
| **info()** | "How is the data structured?" | Check data types, find missing values |
| **describe()** | "What are the numeric patterns?" | Understand distributions, spot outliers |

### Recommended Inspection Workflow:

```python
# Step 1: Load data
df = pd.read_csv('data.csv')

# Step 2: Check shape
print(df.shape)

# Step 3: Preview data
df.head()

# Step 4: Check structure and types
df.info()

# Step 5: Get statistical summary
df.describe()
```

**Always inspect BEFORE cleaning, analysis, or modeling!**

## Complete Inspection Example

Let's demonstrate a complete inspection workflow:

````python
# COMPLETE INSPECTION WORKFLOW

def inspect_dataframe(df, name="DataFrame"):
    """
    Comprehensive DataFrame inspection function.
    
    Parameters:
    -----------
    df : pandas.DataFrame
        The DataFrame to inspect
    name : str
        Name to display in the report
    """
    print("\n" + "=" * 80)
    print(f"INSPECTION REPORT: {name}")
    print("=" * 80)
    
    # 1. Shape
    print(f"\n📊 SHAPE")
    print(f"   Rows: {df.shape[0]:,}")
    print(f"   Columns: {df.shape[1]}")
    
    # 2. Preview
    print(f"\n👀 PREVIEW (First 3 rows)")
    print(df.head(3))
    
    # 3. Structure
    print(f"\n🏗️  STRUCTURE")
    df.info()
    
    # 4. Missing Values
    print(f"\n❓ MISSING VALUES")
    missing = df.isnull().sum()
    missing_percent = (missing / len(df)) * 100
    missing_df = pd.DataFrame({
        'Missing Count': missing,
        'Percentage': missing_percent.round(2)
    })
    if missing.sum() == 0:
        print("   ✓ No missing values detected!")
    else:
        print(missing_df[missing_df['Missing Count'] > 0])
    
    # 5. Numeric Summary
    print(f"\n📈 NUMERIC SUMMARY")
    print(df.describe())
    
    # 6. Categorical Summary
    categorical_cols = df.select_dtypes(include='object').columns
    if len(categorical_cols) > 0:
        print(f"\n📝 CATEGORICAL SUMMARY")
        print(df.describe(include='object'))
    
    print("\n" + "=" * 80)
    print("✓ Inspection Complete")
    print("=" * 80)

# Run the inspection
inspect_dataframe(df, "Employee Survey Q1 2026")
````

## Key Takeaways

### ✅ Always Inspect Before Analysis

1. **head()** - Quick visual preview
   - See column names and sample values
   - Check data alignment
   - Default shows 5 rows

2. **info()** - Structural health check (MOST IMPORTANT)
   - Reveals data types
   - Shows missing values (non-null counts)
   - Displays memory usage

3. **describe()** - Statistical summary
   - Mean, median, std, min, max
   - Percentiles for understanding distribution
   - Helps spot outliers

### 🎯 Best Practices

- **Always run all three methods** when first loading data
- **Check for missing values** using info() and isnull()
- **Verify data types** match expectations
- **Look for outliers** in describe() output
- **Understand your data** before making any transformations

### ⚠️ Common Mistakes to Avoid

- Starting analysis without inspection
- Assuming data types are correct
- Missing hidden null values
- Not checking for outliers
- Relying only on head() for understanding data
