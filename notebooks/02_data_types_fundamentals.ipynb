## 1. Working with Numeric Data Types

Python has two main numeric types:
- **int** (integer): Whole numbers like 42, -17, 0
- **float** (floating-point): Decimal numbers like 3.14, -0.5, 2.0

```python
# Integer Examples
employees_count = 50
survey_responses = 42
departments = 5

print("Integer Examples:")
print(f"Number of employees: {employees_count}")
print(f"Survey responses: {survey_responses}")
print(f"Departments: {departments}")
print(f"Type of employees_count: {type(employees_count)}")
```

```python
# Float Examples
average_satisfaction = 4.2
response_rate = 84.5
average_tenure = 3.75

print("Float Examples:")
print(f"Average satisfaction score: {average_satisfaction}")
print(f"Response rate: {response_rate}%")
print(f"Average tenure: {average_tenure} years")
print(f"Type of average_satisfaction: {type(average_satisfaction)}")
```

### Basic Arithmetic Operations

Python supports all standard arithmetic operations:
- Addition: `+`
- Subtraction: `-`
- Multiplication: `*`
- Division: `/`
- Floor Division: `//`
- Modulo (remainder): `%`
- Exponentiation: `**`

```python
# Basic Arithmetic
total_employees = 100
present_employees = 87

absent_employees = total_employees - present_employees
attendance_percentage = (present_employees / total_employees) * 100

print(f"Total employees: {total_employees}")
print(f"Present: {present_employees}")
print(f"Absent: {absent_employees}")
print(f"Attendance: {attendance_percentage}%")
```

```python
# Division Behavior - Important!
total_hours = 40
work_days = 5

# Regular division always returns a float
hours_per_day = total_hours / work_days
print(f"Regular division: {total_hours} / {work_days} = {hours_per_day} (type: {type(hours_per_day).__name__})")

# Floor division returns an integer (rounded down)
hours_per_day_int = total_hours // work_days
print(f"Floor division: {total_hours} // {work_days} = {hours_per_day_int} (type: {type(hours_per_day_int).__name__})")

# Modulo returns the remainder
remainder = total_hours % 7
print(f"Modulo: {total_hours} % 7 = {remainder}")
```

```python
# Power Operations (Exponentiation)
base_salary = 50000
annual_raise = 1.05  # 5% raise
years = 3

future_salary = base_salary * (annual_raise ** years)

print(f"Starting salary: ${base_salary:,.2f}")
print(f"Annual raise: {(annual_raise - 1) * 100}%")
print(f"Salary after {years} years: ${future_salary:,.2f}")
print(f"Total increase: ${future_salary - base_salary:,.2f}")
```

## 2. Understanding String Data Types

Strings represent text data in Python. They can be created using:
- Single quotes: `'hello'`
- Double quotes: `"hello"`
- Triple quotes: `'''hello'''` or `"""hello"""` (for multi-line strings)

```python
# Creating Strings
employee_name = "Sarah Johnson"
department = 'Engineering'
job_title = """Senior Data Analyst"""

print(f"Employee: {employee_name}")
print(f"Department: {department}")
print(f"Job Title: {job_title}")
print(f"Type: {type(employee_name)}")
```

```python
# String Concatenation (joining strings)
first_name = "John"
last_name = "Doe"

# Using + operator
full_name = first_name + " " + last_name
print(f"Full name: {full_name}")

# Creating greetings
greeting = "Hello, " + full_name + "!"
print(greeting)

# String repetition
divider = "-" * 40
print(divider)
print("Important Message")
print(divider)
```

```python
# Accessing String Characters (Indexing and Slicing)
company = "Trivin Insights"

print(f"Original string: {company}")
print(f"First character: {company[0]}")
print(f"Last character: {company[-1]}")
print(f"First 6 characters: {company[0:6]}")
print(f"Characters from index 7: {company[7:]}")
print(f"Last 8 characters: {company[-8:]}")
print(f"Length: {len(company)} characters")
```

```python
# String Methods
message = "  Python Data Types  "

print(f"Original: '{message}'")
print(f"Uppercase: '{message.upper()}'")
print(f"Lowercase: '{message.lower()}'")
print(f"Stripped: '{message.strip()}'")
print(f"Title case: '{message.title()}'")
print(f"Replaced: '{message.replace('Python', 'Advanced')}'")

# Checking string content
print(f"\nStarts with '  Py': {message.startswith('  Py')}")
print(f"Ends with 'pes  ': {message.endswith('pes  ')}")
print(f"Contains 'Data': {'Data' in message}")
```

```python
# String Formatting (Three methods)
name = "Alice"
score = 87.5
rank = 3

# Method 1: f-strings (recommended - most readable)
result1 = f"{name} scored {score}% and ranked #{rank}"
print(f"F-string: {result1}")

# Method 2: .format() method
result2 = "{} scored {}% and ranked #{}".format(name, score, rank)
print(f"Format: {result2}")

# Method 3: % formatting (old style, less common)
result3 = "%s scored %.1f%% and ranked #%d" % (name, score, rank)
print(f"Old-style: {result3}")

# Advanced f-string formatting
salary = 75000.5678
print(f"\nSalary with 2 decimals: ${salary:.2f}")
print(f"Salary with commas: ${salary:,.2f}")
print(f"Right-aligned (20 chars): {name:>20}")
print(f"Left-aligned (20 chars): {name:<20}|")
```

```python
# Multi-line Strings
report = """
╔════════════════════════════════════════╗
║   EMPLOYEE SURVEY SUMMARY - Q1 2026   ║
╚════════════════════════════════════════╝

Key Metrics:
  • Total Responses: 42
  • Average Rating: 4.2/5.0
  • Completion Rate: 84%
  • Department: Engineering

Status: ✓ Complete
"""

print(report)
```

## 3. Mixing Numbers and Strings Safely

**Critical Concept:** Python does NOT automatically convert between strings and numbers!

This prevents ambiguity but requires explicit conversion.

```python
# Common Error: Mixing Types
age = 30
print(f"Age: {age} (type: {type(age).__name__})")

# This will cause an error!
# Uncomment to see the error:
# message = "Age is " + age  # TypeError: can only concatenate str (not "int") to str

print("❌ Cannot do: 'Age is ' + 30")
print("   Error: TypeError - cannot concatenate string and integer")
```

```python
# Solution 1: Convert number to string using str()
age = 30
message = "Age is " + str(age)
print(f"✓ Correct: {message}")

# Solution 2: Use f-strings (they handle conversion automatically)
message2 = f"Age is {age}"
print(f"✓ Better: {message2}")
```

```python
# Converting Strings to Numbers
text_number = "42"
text_float = "3.14"

print(f"String: '{text_number}' (type: {type(text_number).__name__})")

# Convert to integer
number = int(text_number)
print(f"Integer: {number} (type: {type(number).__name__})")

# Convert to float
decimal = float(text_float)
print(f"Float: {decimal} (type: {type(decimal).__name__})")

# Now we can do arithmetic
result = number + 10
print(f"\nArithmetic: {number} + 10 = {result}")
```

```python
# IMPORTANT: String concatenation vs Addition
score1_str = "85"
score2_str = "92"

# String concatenation (NOT what we want!)
wrong_result = score1_str + score2_str
print(f"String concatenation: '{score1_str}' + '{score2_str}' = '{wrong_result}'")
print(f"Type: {type(wrong_result).__name__}")

# Correct: Convert to numbers first
correct_result = int(score1_str) + int(score2_str)
print(f"\nNumeric addition: {int(score1_str)} + {int(score2_str)} = {correct_result}")
print(f"Type: {type(correct_result).__name__}")
```

```python
# Handling Conversion Errors
valid_number = "123"
invalid_number = "abc"

# This works
result = int(valid_number)
print(f"✓ Converting '{valid_number}' to int: {result}")

# This will fail - but we can handle it
print(f"\nTrying to convert '{invalid_number}' to int...")
try:
    result = int(invalid_number)
except ValueError as e:
    print(f"❌ Error: {e}")
    print("✓ Always validate input before conversion!")
```

## 4. Inspecting Data Types

Python provides tools to check and validate data types:
- `type()`: Returns the type of a variable
- `isinstance()`: Checks if a variable is of a specific type

```python
# Using type() Function
variables = {
    "integer": 42,
    "float": 3.14,
    "string": "Hello",
    "boolean": True,
    "list": [1, 2, 3],
    "dict": {"key": "value"},
    "none": None
}

print("Data Type Inspection:")
print("-" * 60)
for var_name, var_value in variables.items():
    type_name = type(var_value).__name__
    print(f"{var_name:15} = {str(var_value):20} → {type_name}")
```

```python
# Using isinstance() for Type Checking
test_values = [42, 3.14, "text", True, [1, 2], None]

print("Type Checking with isinstance()")
print("-" * 70)
print(f"{'Value':<15} {'int':<8} {'float':<8} {'str':<8} {'bool':<8}")
print("-" * 70)

for value in test_values:
    is_int = isinstance(value, int) and not isinstance(value, bool)
    is_float = isinstance(value, float)
    is_str = isinstance(value, str)
    is_bool = isinstance(value, bool)
    
    print(f"{str(value):<15} {str(is_int):<8} {str(is_float):<8} {str(is_str):<8} {str(is_bool):<8}")
```

```python
# Practical Type Validation
def calculate_bonus(salary, bonus_rate):
    """
    Calculate employee bonus with type validation.
    
    Args:
        salary: Employee salary (numeric)
        bonus_rate: Bonus percentage as decimal (numeric)
    
    Returns:
        Formatted bonus message or error message
    """
    # Validate input types
    if not isinstance(salary, (int, float)):
        return f"❌ Error: Salary must be numeric, got {type(salary).__name__}"
    
    if not isinstance(bonus_rate, (int, float)):
        return f"❌ Error: Bonus rate must be numeric, got {type(bonus_rate).__name__}"
    
    # Calculate bonus
    bonus = salary * bonus_rate
    return f"✓ Bonus: ${bonus:,.2f}"

# Test with correct types
print("Test 1 (correct):")
print(calculate_bonus(50000, 0.10))

# Test with incorrect types
print("\nTest 2 (wrong salary type):")
print(calculate_bonus("50000", 0.10))

print("\nTest 3 (wrong bonus_rate type):")
print(calculate_bonus(50000, "10%"))
```

```python
# Distinguishing Numeric vs String
values = [100, "100", 3.14, "3.14", True, "True"]

print("Distinguishing Numeric from String Values:")
print("-" * 70)
print(f"{'Value':<12} {'Type':<10} {'Is Numeric':<15} {'Is String'}")
print("-" * 70)

for val in values:
    type_name = type(val).__name__
    is_numeric = isinstance(val, (int, float)) and not isinstance(val, bool)
    is_string = isinstance(val, str)
    
    print(f"{str(val):<12} {type_name:<10} {str(is_numeric):<15} {str(is_string)}")
```

## 📊 Summary and Key Takeaways

### What We Learned:

#### 1. **Numeric Types**
- `int`: Whole numbers (42, -17, 0)
- `float`: Decimal numbers (3.14, -0.5)
- Division `/` always returns float
- Floor division `//` returns integer

#### 2. **String Types**
- Created with quotes: `'text'`, `"text"`, `"""text"""`
- Concatenation with `+`
- Indexing and slicing: `text[0]`, `text[0:5]`
- Useful methods: `.upper()`, `.lower()`, `.strip()`, `.replace()`

#### 3. **Type Mixing**
- Cannot mix strings and numbers without conversion
- Use `str()` to convert numbers to strings
- Use `int()` or `float()` to convert strings to numbers
- F-strings handle conversion automatically

#### 4. **Type Inspection**
- `type()` returns the type of a variable
- `isinstance()` checks if variable is a specific type
- Always validate data types before operations

### Best Practices:

✅ **DO:**
- Use f-strings for string formatting
- Validate data types before processing
- Convert types explicitly when needed
- Use meaningful variable names
- Choose appropriate numeric type (int vs float)

❌ **DON'T:**
- Mix types without conversion
- Assume user input is the correct type
- Ignore type errors
- Use string concatenation for arithmetic

### Common Pitfalls:

1. **String + Number**: `"Age: " + 30` → TypeError
2. **String Concatenation vs Addition**: `"5" + "3"` = `"53"` not `8`
3. **Integer Division**: `10 / 3` = `3.333...` not `3`
4. **Boolean is Integer**: `isinstance(True, int)` = `True` (!)
