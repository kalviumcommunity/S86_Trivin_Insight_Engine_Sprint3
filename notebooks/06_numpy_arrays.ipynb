## Section 1: Understanding NumPy Arrays

NumPy (Numerical Python) is a library for numerical computing. At its core is the **ndarray** (n-dimensional array) - a powerful data structure for working with numbers.

### Key Concept: Arrays vs Lists

**Python Lists:**
- Can hold mixed data types (strings, numbers, booleans, etc.)
- Flexible but slower
- Each element is a separate object in memory
- Created with square brackets: `[1, 2, 3]`

**NumPy Arrays:**
- Must hold the same data type (homogeneous)
- Much faster for numeric operations
- Elements stored in contiguous memory
- Created with `np.array([1, 2, 3])`

Let's import NumPy first:


````python
import numpy as np

print("NumPy version:", np.__version__)
````

### Example 1: Python Lists vs NumPy Arrays


````python
# Python List - can hold mixed types
python_list = [1, 2.5, "hello", True]
print("Python List:")
print(f"  Content: {python_list}")
print(f"  Type: {type(python_list)}")
print(f"  Can hold different types: Yes ✓")

# NumPy Array - must be uniform type
numpy_array = np.array([1, 2, 3, 4, 5])
print("\nNumPy Array:")
print(f"  Content: {numpy_array}")
print(f"  Type: {type(numpy_array)}")
print(f"  Data type of elements: {numpy_array.dtype}")
print(f"  Must be uniform type: Yes ✓")
````

### Example 2: What Happens With Mixed Types in an Array?


````python
# When you try to create an array with mixed types, NumPy converts to a common type
mixed_array = np.array([1, 2.5, 3])
print("Array from [1, 2.5, 3]:")
print(f"  Content: {mixed_array}")
print(f"  Data type: {mixed_array.dtype}")
print(f"  → NumPy converted everything to float64!\n")

mixed_with_string = np.array([1, 2, "hello"])
print("Array from [1, 2, 'hello']:")
print(f"  Content: {mixed_with_string}")
print(f"  Data type: {mixed_with_string.dtype}")
print(f"  → NumPy converted everything to string (object)!")
````

## Section 2: Creating NumPy Arrays from Lists

The most common way to create NumPy arrays is from Python lists using `np.array()`. You can create:
- **1D arrays** from simple lists
- **2D arrays** (matrices) from nested lists
- **3D+ arrays** from deeper nested lists

### Creating 1D Arrays


````python
# Create a 1D array from a list
numbers_list = [10, 20, 30, 40, 50]
numbers_array = np.array(numbers_list)

print("Creating a 1D Array:")
print(f"  Python list: {numbers_list}")
print(f"  NumPy array: {numbers_array}")
print(f"  Array display:\n{numbers_array}")
print(f"  Type: {type(numbers_array)}")
print(f"  Data type: {numbers_array.dtype}")
````

### Creating 2D Arrays (Matrices)


````python
# Create a 2D array from a nested list
matrix_list = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

matrix_array = np.array(matrix_list)

print("Creating a 2D Array (Matrix):")
print(f"  Python nested list:\n  {matrix_list}\n")
print(f"  NumPy 2D array:")
print(matrix_array)
print(f"\n  Type: {type(matrix_array)}")
print(f"  Data type: {matrix_array.dtype}")
print(f"  This is a grid with 3 rows and 3 columns")
````

### Specifying Data Types Explicitly


````python
# NumPy infers data type automatically
auto_int = np.array([1, 2, 3, 4])
print("Automatic type detection:")
print(f"  [1, 2, 3, 4] → dtype: {auto_int.dtype}")

auto_float = np.array([1.0, 2.0, 3.0, 4.0])
print(f"  [1.0, 2.0, 3.0, 4.0] → dtype: {auto_float.dtype}\n")

# You can specify the dtype explicitly with the dtype parameter
print("Explicit data type specification:")
explicit_int = np.array([1, 2, 3], dtype=int)
print(f"  dtype=int → {explicit_int.dtype}: {explicit_int}")

explicit_float = np.array([1, 2, 3], dtype=float)
print(f"  dtype=float → {explicit_float.dtype}: {explicit_float}")

explicit_complex = np.array([1, 2, 3], dtype=complex)
print(f"  dtype=complex → {explicit_complex.dtype}: {explicit_complex}")
````

## Section 3: Inspecting Array Properties

Understanding array properties is essential for debugging and ensuring your code works correctly. Key properties include:
- **shape**: Dimensions of the array (rows, columns, etc.)
- **dtype**: Data type of elements
- **ndim**: Number of dimensions
- **size**: Total number of elements
- **itemsize**: Size of each element in bytes

### Inspecting 1D Arrays


````python
# Create a 1D array
array_1d = np.array([10, 20, 30, 40, 50])

print("Properties of 1D Array:")
print(f"  Array: {array_1d}")
print(f"  shape: {array_1d.shape}       → (5,) means 5 elements in 1 dimension")
print(f"  dtype: {array_1d.dtype}       → int64 is 64-bit integer")
print(f"  ndim: {array_1d.ndim}         → 1 dimension")
print(f"  size: {array_1d.size}         → 5 total elements")
print(f"  itemsize: {array_1d.itemsize}       → {array_1d.itemsize} bytes per element")
````

### Inspecting 2D Arrays


````python
# Create a 2D array
array_2d = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Properties of 2D Array:")
print(f"  Array:\n{array_2d}\n")
print(f"  shape: {array_2d.shape}       → (2, 3) means 2 rows, 3 columns")
print(f"  dtype: {array_2d.dtype}")
print(f"  ndim: {array_2d.ndim}         → 2 dimensions")
print(f"  size: {array_2d.size}         → 6 total elements (2 × 3)")
print(f"  itemsize: {array_2d.itemsize}       → {array_2d.itemsize} bytes per element\n")

# A real-world example: Employee survey data
print("Real-world example: Employee Survey")
survey_data = np.array([
    [8, 7, 9],  # Employee 1 scores
    [6, 8, 7],  # Employee 2 scores
    [7, 9, 8]   # Employee 3 scores
])
print(f"  Survey scores (3 employees, 3 questions):")
print(f"{survey_data}")
print(f"  shape: {survey_data.shape} → 3 employees, 3 survey questions")
````

## Section 4: Basic Operations on Arrays

NumPy arrays support element-wise arithmetic operations. This means each operation is applied to corresponding elements.

### Element-Wise Arithmetic Operations


````python
# Element-wise arithmetic
array1 = np.array([10, 20, 30, 40])
array2 = np.array([1, 2, 3, 4])

print("Element-wise Arithmetic Operations:")
print(f"array1: {array1}")
print(f"array2: {array2}\n")

print(f"Addition (+):\n  {array1} + {array2} = {array1 + array2}")
print(f"\nSubtraction (-):\n  {array1} - {array2} = {array1 - array2}")
print(f"\nMultiplication (*):\n  {array1} * {array2} = {array1 * array2}")
print(f"\nDivision (/):\n  {array1} / {array2} = {array1 / array2}")
print(f"\nPower (**):\n  {array2} ** 2 = {array2 ** 2}")
````

### CRITICAL DIFFERENCE: Lists vs Arrays in Operations


````python
# WATCH OUT! This is a common source of confusion!

print("With Python LISTS:")
print("=" * 50)
list1 = [1, 2, 3]
list2 = [4, 5, 6]
result_list = list1 + list2
print(f"{list1} + {list2} = {result_list}")
print(f"Result: Lists are CONCATENATED (joined together)!")

print("\n\nWith NumPy ARRAYS:")
print("=" * 50)
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
result_array = array1 + array2
print(f"{array1} + {array2} = {result_array}")
print(f"Result: Arrays are added ELEMENT-WISE!")

print("\n\nWhy this matters:")
print("-" * 50)
print("✗ WRONG: You cannot use lists for math")
print("✓ RIGHT: Use NumPy arrays for mathematical operations")
````

### Scalar Operations (Broadcasting)


````python
# Operations with a single value (scalar) and array
array = np.array([10, 20, 30, 40, 50])

print("Broadcasting: Applying a scalar operation to an entire array")
print(f"Original array: {array}\n")

print(f"Add 5 to each element:    {array + 5}")
print(f"Subtract 3 from each:     {array - 3}")
print(f"Multiply each by 2:       {array * 2}")
print(f"Divide each by 10:        {array / 10}")
print(f"Square each element:      {array ** 2}")

print("\n" + "=" * 60)
print("WHY THIS IS POWERFUL: Multi-dimensional arrays")
print("=" * 60)

matrix = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print(f"Original matrix:\n{matrix}\n")
print(f"Multiply entire matrix by 3:\n{matrix * 3}\n")
print(f"Scale scores from 0-10 to 0-100:\n{matrix * 10}")
````

### Useful Mathematical Functions


````python
# Common mathematical functions applied element-wise
array = np.array([1, 4, 9, 16, 25])

print("Mathematical Functions (applied element-wise):")
print(f"Original array: {array}\n")
print(f"Square root: {np.sqrt(array)}")
print(f"Absolute value: {np.abs(np.array([-1, -2, 3, -4]))}")
print(f"Exponential (e^x): {np.exp([0, 1, 2])}")
print(f"Natural log: {np.log([1, np.e, np.e**2])}")
print(f"Sine: {np.sin([0, np.pi/2, np.pi])}")
print(f"Cosine: {np.cos([0, np.pi/2, np.pi])}")
````

### Aggregate Functions (Summarizing Data)


````python
# Aggregate functions summarize all data into a single value
satisfaction_scores = np.array([7, 8, 6, 9, 8, 7])

print("Aggregate Functions on 1D array:")
print(f"Score array: {satisfaction_scores}\n")
print(f"Sum:         {np.sum(satisfaction_scores)}")
print(f"Mean:        {np.mean(satisfaction_scores):.2f}")
print(f"Median:      {np.median(satisfaction_scores):.2f}")
print(f"Min:         {np.min(satisfaction_scores)}")
print(f"Max:         {np.max(satisfaction_scores)}")
print(f"Std Dev:     {np.std(satisfaction_scores):.2f}  (spread of data)")

print("\n" + "=" * 60)
print("2D Array: Department satisfaction scores")
print("=" * 60)

dept_scores = np.array([
    [7, 8, 6, 9],      # Department A
    [9, 8, 9, 8],      # Department B
    [6, 7, 6, 5]       # Department C
])

print(f"Scores by department (rows):\n{dept_scores}\n")
print(f"Overall mean: {np.mean(dept_scores):.2f}")
print(f"Mean by department: {np.mean(dept_scores, axis=1)}")
print(f"Mean by question: {np.mean(dept_scores, axis=0)}")
````

## Summary & Key Takeaways

### What We Learned

1. **NumPy Arrays vs Python Lists**
   - Arrays are **homogeneous** (all same type), lists are heterogeneous
   - Arrays are **faster** and more **memory-efficient**
   - Use arrays for numerical computing

2. **Creating Arrays**
   - `np.array()` converts lists to arrays
   - Can create 1D, 2D, 3D, and higher dimensional arrays
   - Explicitly specify `dtype` when needed

3. **Array Properties**
   - **shape**: dimensions (rows, columns, etc.)
   - **dtype**: data type of elements
   - **ndim**: number of dimensions
   - **size**: total number of elements

4. **Operations on Arrays**
   - **Element-wise**: `array1 + array2` adds corresponding elements
   - **Broadcasting**: operations with scalars apply to all elements
   - **Aggregate functions**: sum, mean, min, max, std summarize data

5. **Critical Difference from Lists**
   - Lists: `[1,2,3] + [1,2,3] = [1,2,3,1,2,3]` (concatenation)
   - Arrays: `array([1,2,3]) + array([1,2,3]) = array([2,4,6])` (addition)
