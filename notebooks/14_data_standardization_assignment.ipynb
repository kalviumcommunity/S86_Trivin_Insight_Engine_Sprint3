## Task 2: Standardize Column Names (20 points)

Convert all column names to snake_case format.

**Requirements:**
- Convert to lowercase
- Replace spaces with underscores
- Remove special characters
- Apply consistent naming

```python
# TODO: Create a function to standardize column names

def standardize_column_names(df):
    """
    Standardize column names to snake_case format.
    
    Parameters:
    -----------
    df : pd.DataFrame
        Input DataFrame
        
    Returns:
    --------
    pd.DataFrame
        DataFrame with standardized column names
    """
    df_clean = df.copy()
    new_columns = []
    
    for col in df_clean.columns:
        # Convert to lowercase
        col_clean = col.lower()
        
        # Replace spaces with underscores
        col_clean = col_clean.replace(' ', '_')
        
        # Remove special characters (keep only alphanumeric and underscore)
        col_clean = re.sub(r'[^a-z0-9_]', '', col_clean)
        
        # Remove multiple consecutive underscores
        col_clean = re.sub(r'_+', '_', col_clean)
        
        # Remove leading/trailing underscores
        col_clean = col_clean.strip('_')
        
        new_columns.append(col_clean)
    
    df_clean.columns = new_columns
    return df_clean

print("✓ Standardization function created")
```

```python
# TODO: Apply the standardization function
df_standardized = standardize_column_names(df)

print("Column Name Transformation:")
for orig, new in zip(df.columns, df_standardized.columns):
    if orig != new:
        print(f"  '{orig}' → '{new}'")
    else:
        print(f"  '{orig}' (no change needed)")
```

```python
# TODO: Display the standardized column names
print("Standardized Column Names:")
for i, col in enumerate(df_standardized.columns, 1):
    print(f"{i}. {col}")
```

## Task 3: Standardize Data Formats (20 points)

Clean and normalize the actual data values.

### 3.1 Standardize Text Columns

```python
# TODO: Identify text columns that need standardization
text_cols = [col for col in df_standardized.columns 
             if 'department' in col or 'name' in col]

print(f"Text columns to standardize: {text_cols}")

# Show current state
if text_cols:
    print("\nCurrent values:")
    for col in text_cols:
        if col in df_standardized.columns:
            print(f"\n{col}:")
            print(df_standardized[col].unique())
```

```python
# TODO: Standardize text data
# Strip whitespace and apply consistent casing

df_clean = df_standardized.copy()

for col in text_cols:
    if col in df_clean.columns:
        # Strip leading/trailing whitespace
        df_clean[col] = df_clean[col].astype(str).str.strip()
        
        # Apply title case for consistency
        df_clean[col] = df_clean[col].str.title()

print("✓ Text columns standardized")

# Show after standardization
if text_cols:
    print("\nStandardized values:")
    for col in text_cols:
        if col in df_clean.columns:
            print(f"\n{col}:")
            print(df_clean[col].unique())
```

### 3.2 Standardize Numeric Columns

```python
# TODO: Identify numeric columns
numeric_cols = [col for col in df_clean.columns 
                if any(x in col for x in ['score', 'balance', 'support', 'growth', 'collaboration'])]

print(f"Numeric columns: {numeric_cols}")

# Ensure they are numeric type
for col in numeric_cols:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

print("✓ Numeric columns verified")
```

### 3.3 Standardize Date Columns

```python
# TODO: Identify and convert date columns
date_cols = [col for col in df_clean.columns if 'date' in col]

print(f"Date columns: {date_cols}")

for col in date_cols:
    if col in df_clean.columns:
        df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')
        print(f"  {col}: converted to datetime")

print("\n✓ Date columns standardized")
```

## Task 4: Identify and Remove Duplicate Records (20 points)

Check for duplicate rows and remove them if found.

```python
# TODO: Check for duplicates
print("Checking for duplicate records...\n")

# Count duplicates
duplicate_count = df_clean.duplicated().sum()
print(f"Total duplicate rows: {duplicate_count}")

if duplicate_count > 0:
    print("\n⚠️ Duplicate rows found:")
    print(df_clean[df_clean.duplicated(keep=False)])
else:
    print("\n✓ No duplicate records found")
```

```python
# TODO: Remove duplicates (if any)
rows_before = len(df_clean)

df_clean = df_clean.drop_duplicates(keep='first')

rows_after = len(df_clean)
removed = rows_before - rows_after

print(f"Rows before: {rows_before}")
print(f"Rows after:  {rows_after}")
print(f"Removed:     {removed}")

if removed > 0:
    print(f"\n✓ Removed {removed} duplicate record(s)")
else:
    print("\n✓ No duplicates to remove")
```

Calculate and display summary statistics for individual numeric columns.

Generate a comprehensive data quality report.
