## 1. Working with Python Lists

**Lists** are:
- **Ordered**: Elements maintain their position
- **Mutable**: Can be changed after creation
- **Indexed by position**: Access elements using numbers (0, 1, 2...)
- **Allow duplicates**: Same value can appear multiple times

**Syntax**: `[item1, item2, item3]`

```python
# Creating Lists
employees = ["Alice", "Bob", "Charlie", "Diana"]
scores = [85, 92, 78, 95, 88]
mixed_data = ["John", 30, 75000.50, True]
empty_list = []

print(f"Employees: {employees}")
print(f"Scores: {scores}")
print(f"Mixed data: {mixed_data}")
print(f"Empty list: {empty_list}")
print(f"Type: {type(employees)}")
```

```python
# Accessing List Elements (Indexing)
departments = ["Engineering", "Sales", "Marketing", "HR", "Finance"]

print(f"All departments: {departments}")
print(f"First (index 0): {departments[0]}")
print(f"Third (index 2): {departments[2]}")
print(f"Last (index -1): {departments[-1]}")
print(f"Second to last: {departments[-2]}")
print(f"\nTotal departments: {len(departments)}")
```

```python
# List Slicing
numbers = [10, 20, 30, 40, 50, 60, 70]

print(f"Original: {numbers}")
print(f"First 3 elements [0:3]: {numbers[0:3]}")
print(f"First 3 elements [:3]: {numbers[:3]}")
print(f"From index 3 onwards [3:]: {numbers[3:]}")
print(f"From index 2 to 5 [2:5]: {numbers[2:5]}")
print(f"Last 3 elements [-3:]: {numbers[-3:]}")
print(f"Every 2nd element [::2]: {numbers[::2]}")
```

```python
# Modifying Lists (Lists are MUTABLE)
survey_scores = [4, 5, 3, 4, 5]
print(f"Original: {survey_scores}")

# Change an element
survey_scores[2] = 5
print(f"After changing index 2 to 5: {survey_scores}")

# Add to end
survey_scores.append(4)
print(f"After append(4): {survey_scores}")

# Insert at specific position
survey_scores.insert(0, 3)
print(f"After insert(0, 3): {survey_scores}")

# Remove first occurrence
survey_scores.remove(3)
print(f"After remove(3): {survey_scores}")

# Remove and return last element
last = survey_scores.pop()
print(f"Popped: {last}, Remaining: {survey_scores}")
```

```python
# List Methods
ratings = [4.2, 4.5, 3.8, 4.9, 4.1]

print(f"Original: {ratings}")
print(f"Minimum: {min(ratings)}")
print(f"Maximum: {max(ratings)}")
print(f"Sum: {sum(ratings)}")
print(f"Average: {sum(ratings) / len(ratings):.2f}")

# Sort (modifies in place)
ratings.sort()
print(f"Sorted: {ratings}")

# Reverse
ratings.reverse()
print(f"Reversed: {ratings}")
```

```python
# List Operations
team_a = ["Alice", "Bob"]
team_b = ["Charlie", "Diana"]

# Concatenation
all_members = team_a + team_b
print(f"Combined: {all_members}")

# Repetition
repeated = ["X"] * 5
print(f"Repeated: {repeated}")

# Membership
print(f"'Alice' in team_a? {'Alice' in team_a}")
print(f"'Eve' in team_a? {'Eve' in team_a}")
```

## 2. Working with Python Tuples

**Tuples** are:
- **Ordered**: Elements maintain their position
- **Immutable**: Cannot be changed after creation
- **Indexed by position**: Access elements using numbers (0, 1, 2...)
- **Allow duplicates**: Same value can appear multiple times

**Syntax**: `(item1, item2, item3)`

```python
# Creating Tuples
coordinates = (10, 20)
employee_record = ("John Doe", 1001, "Engineering", 75000)
single_item = (42,)  # Note the comma for single element!
empty_tuple = ()

print(f"Coordinates: {coordinates}")
print(f"Employee: {employee_record}")
print(f"Single item: {single_item}")
print(f"Empty: {empty_tuple}")
print(f"Type: {type(coordinates)}")
```

```python
# Accessing Tuple Elements
person = ("Alice", 30, "Data Scientist", "New York")

print(f"Full tuple: {person}")
print(f"Name (index 0): {person[0]}")
print(f"Age (index 1): {person[1]}")
print(f"Job (index 2): {person[2]}")
print(f"City (index -1): {person[-1]}")
print(f"First two: {person[:2]}")
```

```python
# Tuple Unpacking (Very Useful!)
employee = ("Bob Smith", "Engineering", 85000)

# Unpack into variables
name, department, salary = employee

print(f"Name: {name}")
print(f"Department: {department}")
print(f"Salary: ${salary:,.2f}")

# Swapping values using tuples
x, y = 10, 20
print(f"\nBefore swap: x={x}, y={y}")
x, y = y, x  # Elegant swap!
print(f"After swap: x={x}, y={y}")
```

```python
# Tuple Immutability
config = ("localhost", 8080, "production")
print(f"Original config: {config}")

# This would cause an error:
# config[1] = 9090  # TypeError: 'tuple' object does not support item assignment

print("\n❌ Cannot modify: config[1] = 9090")
print("✓ Tuples are immutable - this protects data")

# To 'change' a tuple, create a new one
new_config = (config[0], 9090, config[2])
print(f"\nNew tuple created: {new_config}")
```

```python
# When to Use Tuples
print("Use tuples when:")
print("  ✓ Data should NOT change (coordinates, RGB colors)")
print("  ✓ Returning multiple values from a function")
print("  ✓ Dictionary keys (must be immutable)")
print("  ✓ Performance matters (tuples are faster than lists)")

# Example: Function returning multiple values
def get_stats(scores):
    """Return min, max, average as tuple"""
    return min(scores), max(scores), sum(scores) / len(scores)

minimum, maximum, average = get_stats([85, 92, 78, 95, 88])
print(f"\nStats: min={minimum}, max={maximum}, avg={average:.1f}")
```

## 3. Working with Python Dictionaries

**Dictionaries** are:
- **Unordered** (3.7+: maintains insertion order)
- **Mutable**: Can be changed after creation
- **Indexed by key**: Access values using meaningful keys
- **Keys are unique**: Each key appears only once

**Syntax**: `{key1: value1, key2: value2}`

```python
# Creating Dictionaries
employee = {
    "name": "Alice Johnson",
    "id": 1001,
    "department": "Data Science",
    "salary": 75000
}

scores = {"math": 85, "science": 92, "english": 88}
empty_dict = {}

print(f"Employee: {employee}")
print(f"Scores: {scores}")
print(f"Empty: {empty_dict}")
print(f"Type: {type(employee)}")
```

```python
# Accessing Dictionary Values
employee = {
    "name": "Bob Smith",
    "id": 1002,
    "department": "Engineering",
    "salary": 85000
}

# Access with square brackets
print(f"Name: {employee['name']}")
print(f"Department: {employee['department']}")

# Safe access with .get()
print(f"\nSalary: {employee.get('salary')}")
print(f"Age (doesn't exist): {employee.get('age')}")
print(f"Age with default: {employee.get('age', 'Not specified')}")
```

```python
# Modifying Dictionaries (Dictionaries are MUTABLE)
survey = {
    "id": 101,
    "score": 4.5,
    "department": "Sales"
}

print(f"Original: {survey}")

# Update existing key
survey["score"] = 4.8
print(f"After updating score: {survey}")

# Add new key-value pair
survey["date"] = "2026-02-27"
print(f"After adding date: {survey}")

# Remove key-value pair
removed = survey.pop("department")
print(f"Removed: {removed}")
print(f"After removal: {survey}")
```

```python
# Dictionary Methods
data = {"a": 1, "b": 2, "c": 3}

print(f"Dictionary: {data}")
print(f"Keys: {list(data.keys())}")
print(f"Values: {list(data.values())}")
print(f"Items: {list(data.items())}")

# Check if key exists
print(f"\n'a' in data? {'a' in data}")
print(f"'z' in data? {'z' in data}")
```

```python
# Iterating Over Dictionaries
dept_counts = {
    "Engineering": 25,
    "Sales": 18,
    "Marketing": 12,
    "HR": 8
}

print("Department Employee Counts:")
for department, count in dept_counts.items():
    print(f"  {department}: {count} employees")

# Calculate total
total = sum(dept_counts.values())
print(f"\nTotal employees: {total}")
```

```python
# Nested Dictionaries
employees = {
    "emp001": {
        "name": "Alice",
        "department": "Engineering",
        "salary": 75000
    },
    "emp002": {
        "name": "Bob",
        "department": "Sales",
        "salary": 65000
    }
}

print("Employee Database:")
for emp_id, emp_data in employees.items():
    print(f"  {emp_id}: {emp_data['name']} - ${emp_data['salary']:,}")

# Access nested data
alice_dept = employees["emp001"]["department"]
print(f"\nAlice's department: {alice_dept}")
```

## 4. Choosing the Right Data Structure

Understanding when to use each collection type is crucial for writing effective code.

```python
# Comparison Table
import pandas as pd

comparison = pd.DataFrame({
    'Feature': ['Ordered', 'Mutable', 'Indexed by', 'Duplicates', 'Syntax'],
    'List': ['Yes', 'Yes', 'Position (0,1,2...)', 'Yes', '[1, 2, 3]'],
    'Tuple': ['Yes', 'No', 'Position (0,1,2...)', 'Yes', '(1, 2, 3)'],
    'Dictionary': ['Yes (3.7+)', 'Yes', 'Key (any hashable)', 'Keys: No, Values: Yes', '{"a": 1}']
})

print(comparison.to_string(index=False))
```

```python
# Use Case: Lists
print("✓ Use LISTS when:")
print("  - Order matters")
print("  - You need to modify contents")
print("  - You have similar items")
print("  - You need to sort/filter")

# Example: Daily survey scores
daily_scores = [4.2, 4.5, 4.3, 4.7, 4.4]
print(f"\nExample - Daily scores: {daily_scores}")
print(f"Average: {sum(daily_scores) / len(daily_scores):.2f}")
```

```python
# Use Case: Tuples
print("✓ Use TUPLES when:")
print("  - Data should NOT change")
print("  - Protecting from accidental modification")
print("  - Returning multiple values")
print("  - Using as dictionary keys")

# Example: Geographic coordinates as dictionary keys
locations = {
    (40.7128, -74.0060): "New York",
    (34.0522, -118.2437): "Los Angeles"
}
print(f"\nExample - Location lookup:")
print(f"  {locations[(40.7128, -74.0060)]}")
```

```python
# Use Case: Dictionaries
print("✓ Use DICTIONARIES when:")
print("  - You need key-based lookup")
print("  - Modeling entities with attributes")
print("  - Data has labels/names")
print("  - You need fast access")

# Example: Employee profile
employee = {
    "id": 1001,
    "name": "Alice",
    "department": "Data Science",
    "skills": ["Python", "SQL", "ML"]
}
print(f"\nExample - Employee profile:")
print(f"  Name: {employee['name']}")
print(f"  Skills: {', '.join(employee['skills'])}")
```

```python
# Same Data, Different Structures
print("Representing a person:")

# As list (positional)
person_list = ["John", 30, "Engineer"]
print(f"\nList: {person_list}")
print(f"  Access: person_list[0] = {person_list[0]}")
print("  ❌ Issue: What does index 2 mean?")

# As tuple (positional, immutable)
person_tuple = ("John", 30, "Engineer")
print(f"\nTuple: {person_tuple}")
print(f"  Access: person_tuple[0] = {person_tuple[0]}")
print("  ✓ Benefit: Cannot accidentally modify")

# As dictionary (labeled)
person_dict = {"name": "John", "age": 30, "job": "Engineer"}
print(f"\nDictionary: {person_dict}")
print(f"  Access: person_dict['name'] = {person_dict['name']}")
print("  ✓✓ Benefit: Self-documenting!")
```

## 5. Practical Application: Employee Survey System

Let's combine all three collection types in a real-world scenario.

```python
# Survey questions (List - ordered, can add more)
questions = [
    "Rate work-life balance (1-5)",
    "Rate career growth (1-5)",
    "Rate management (1-5)",
    "Would you recommend?"
]

print("Survey Questions:")
for i, q in enumerate(questions, 1):
    print(f"  Q{i}. {q}")
```

```python
# Employee info (Tuple - shouldn't change)
employee_info = (1001, "Jennifer Smith", "Data Science", "2020-01-15")
emp_id, emp_name, emp_dept, hire_date = employee_info

print(f"Employee: {emp_name}")
print(f"Department: {emp_dept}")
print(f"ID: {emp_id}")
```

```python
# Survey response (Dictionary - labeled data)
response = {
    "employee_id": emp_id,
    "response_date": "2026-02-27",
    "scores": {
        "work_life_balance": 4.5,
        "career_growth": 4.2,
        "management": 4.8
    },
    "would_recommend": True,
    "comments": "Great team culture"
}

print(f"\nSurvey Response:")
print(f"  Date: {response['response_date']}")
print(f"  Scores:")
for category, score in response['scores'].items():
    print(f"    {category}: {score}")
print(f"  Recommends: {response['would_recommend']}")
```

```python
# Multiple responses (List of dictionaries)
all_responses = [
    {"emp_id": 1001, "dept": "Data Science", "score": 4.5, "recommend": True},
    {"emp_id": 1002, "dept": "Engineering", "score": 4.8, "recommend": True},
    {"emp_id": 1003, "dept": "Sales", "score": 3.9, "recommend": False}
]

print("All Responses:")
for r in all_responses:
    print(f"  Emp {r['emp_id']}: Score {r['score']}, Recommend: {r['recommend']}")

# Calculate statistics
avg_score = sum(r['score'] for r in all_responses) / len(all_responses)
recommend_rate = sum(1 for r in all_responses if r['recommend']) / len(all_responses) * 100

print(f"\nStatistics:")
print(f"  Average score: {avg_score:.2f}")
print(f"  Recommendation rate: {recommend_rate:.0f}%")
```

## 📊 Summary and Key Takeaways

### Quick Reference

| Collection | Mutable | Ordered | Syntax | Best For |
|------------|---------|---------|--------|----------|
| **List** | ✅ Yes | ✅ Yes | `[1, 2, 3]` | Dynamic collections |
| **Tuple** | ❌ No | ✅ Yes | `(1, 2, 3)` | Fixed data |
| **Dictionary** | ✅ Yes | ✅ Yes (3.7+) | `{"a": 1}` | Labeled data |

### Key Operations

**Lists:**
- Create: `my_list = [1, 2, 3]`
- Access: `my_list[0]`
- Modify: `my_list[0] = 10`
- Add: `my_list.append(4)`
- Remove: `my_list.remove(2)`

**Tuples:**
- Create: `my_tuple = (1, 2, 3)`
- Access: `my_tuple[0]`
- Unpack: `a, b, c = my_tuple`
- Cannot modify!

**Dictionaries:**
- Create: `my_dict = {"a": 1, "b": 2}`
- Access: `my_dict["a"]` or `my_dict.get("a")`
- Modify: `my_dict["a"] = 10`
- Add: `my_dict["c"] = 3`
- Remove: `my_dict.pop("a")`

### Best Practices

✅ **DO:**
- Use lists for collections that change
- Use tuples to protect data
- Use dictionaries for labeled data
- Choose meaningful names
- Use .get() for safe dictionary access

❌ **DON'T:**
- Mix up when to use each type
- Try to modify tuples
- Use lists when dictionaries are clearer
- Forget the comma in single-item tuples
