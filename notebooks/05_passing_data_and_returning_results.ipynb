````python
# Exercise 2: Grade Calculator System

def validate_score(score):
    """Validate that score is in valid range"""
    return 0 <= score <= 100

def score_to_letter_grade(score):
    """Convert numerical score to letter grade"""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

def score_to_gpa(score):
    """Convert score to GPA (4.0 scale)"""
    if score >= 90:
        return 4.0
    elif score >= 80:
        return 3.5
    elif score >= 70:
        return 3.0
    elif score >= 60:
        return 2.5
    else:
        return 0.0

def calculate_average_score(scores):
    """Calculate average of scores, filtering out invalid ones"""
    valid_scores = [s for s in scores if validate_score(s)]
    if not valid_scores:
        return 0
    return sum(valid_scores) / len(valid_scores)

def grade_report(score):
    """Generate grade report for a single score"""
    if not validate_score(score):
        return f"Invalid score: {score}. Must be 0-100."
    
    letter = score_to_letter_grade(score)
    gpa = score_to_gpa(score)
    
    return f"Score: {score}% → Grade: {letter} → GPA: {gpa:.1f}"

# Test the functions
print("\nEXERCISE 2: Grade Calculator")
print("="*60)

student_scores = [92, 85, 78, 95, 88, 105, 72, -5, 81]
print(f"Student scores: {student_scores}")

# Show individual grade reports
print("\nIndividual Grade Reports:")
for score in student_scores:
    print(f"  {grade_report(score)}")

# Calculate class average
valid_scores = [s for s in student_scores if validate_score(s)]
class_avg = calculate_average_score(student_scores)
print(f"\nClass Statistics:")
print(f"  Valid scores: {len(valid_scores)} out of {len(student_scores)}")
print(f"  Class average: {class_avg:.2f}%")
print(f"  Average grade: {score_to_letter_grade(class_avg)}")
print(f"  Average GPA: {score_to_gpa(class_avg):.2f}")
````

````python
# Section 8: Building Reusable and Composable Functions

# ============================================================================
# Example: Employee Analytics System
# ============================================================================

# TIER 1: Basic calculation functions
def calculate_years_of_service(start_year, current_year=2026):
    """Calculate how many years someone has worked"""
    return current_year - start_year

def validate_employee_salary(salary, min_salary=20000, max_salary=250000):
    """Check if salary is within valid range"""
    return min_salary <= salary <= max_salary

def calculate_performance_bonus(base_salary, performance_rating):
    """Calculate bonus based on performance"""
    rating_to_bonus = {
        "Excellent": 0.20,
        "Good": 0.10,
        "Average": 0.05,
        "Poor": 0.00
    }
    bonus_percent = rating_to_bonus.get(performance_rating, 0)
    return base_salary * bonus_percent

# TIER 2: Analysis functions (use TIER 1 functions)
def analyze_employee_compensation(name, base_salary, performance_rating):
    """Analyze complete compensation package"""
    # Validate input
    is_valid = validate_employee_salary(base_salary)
    if not is_valid:
        return None
    
    # Calculate bonus
    bonus = calculate_performance_bonus(base_salary, performance_rating)
    total_comp = base_salary + bonus
    
    return {
        "name": name,
        "base_salary": base_salary,
        "bonus": bonus,
        "total_compensation": total_comp,
        "rating": performance_rating
    }

def analyze_employee_tenure(name, start_year):
    """Analyze tenure and suggest benefits"""
    years = calculate_years_of_service(start_year)
    
    if years >= 10:
        tenure_status = "Veteran"
    elif years >= 5:
        tenure_status = "Senior"
    else:
        tenure_status = "New/Growing"
    
    return {
        "name": name,
        "years_of_service": years,
        "status": tenure_status
    }

# TIER 3: Reporting functions (use TIER 2 functions)
def generate_employee_report(name, base_salary, performance_rating, start_year):
    """Generate comprehensive employee report"""
    comp_data = analyze_employee_compensation(name, base_salary, performance_rating)
    tenure_data = analyze_employee_tenure(name, start_year)
    
    if comp_data is None:
        return "Invalid salary provided"
    
    report = f"""
    EMPLOYEE REPORT: {name}
    ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    Compensation:
      Base Salary: ${comp_data['base_salary']:,}
      Bonus ({comp_data['rating']}): ${comp_data['bonus']:,.2f}
      Total: ${comp_data['total_compensation']:,.2f}
    
    Tenure:
      Years Service: {tenure_data['years_of_service']}
      Status: {tenure_data['status']}
    ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
    """
    return report

print("REUSABLE AND COMPOSABLE FUNCTIONS:")
print("="*60)

# Use the system
report1 = generate_employee_report(
    name="Priya Kumar",
    base_salary=95000,
    performance_rating="Excellent",
    start_year=2018
)
print(report1)

report2 = generate_employee_report(
    name="James Chen",
    base_salary=75000,
    performance_rating="Good",
    start_year=2023
)
print(report2)

print("="*60)
print("\nBENEFITS OF THIS DESIGN:")
print("="*60)
print("""
✅ Each function has ONE clear purpose
✅ Easy to test individual functions
✅ Easy to modify behavior (e.g., change bonus structure)
✅ Code is highly reusable
✅ Clear data flow: Input → Processing → Output
✅ Can build complex reports from simple functions
✅ Easy to add new features
""")
````

````python
# Section 7: Avoiding Common Function Mistakes

print("MISTAKE 1: Printing Instead of Returning")
print("="*60)

# ❌ BAD: Only prints, can't reuse
def calculate_discount_bad(price, discount):
    result = price * (1 - discount / 100)
    print(f"Discounted price: ${result:.2f}")  # Only prints!

# ✅ GOOD: Returns the value
def calculate_discount_good(price, discount):
    result = price * (1 - discount / 100)
    return result

print("❌ BAD approach (only prints, result is lost):")
calculate_discount_bad(100, 20)
print("Can't use the result for further calculations")

print("\n✅ GOOD approach (returns value, can be reused):")
discounted = calculate_discount_good(100, 20)
with_tax = discounted * 1.08
print(f"Price: $100 → After 20% discount: ${discounted:.2f}")
print(f"After 8% tax: ${with_tax:.2f}")

print("\n" + "="*60)
print("MISTAKE 2: Hardcoded Values")
print("="*60)

# ❌ BAD: Commission rate hardcoded
def calculate_commission_bad(sales):
    return sales * 0.05  # What if rate changes? Hard to modify!

# ✅ GOOD: Commission rate as parameter
def calculate_commission_good(sales, rate):
    return sales * rate

print("❌ BAD (hardcoded 5%):")
print(f"Commission on $1000: ${calculate_commission_bad(1000):.2f}")

print("\n✅ GOOD (rate is flexible):")
print(f"Commission at 5%: ${calculate_commission_good(1000, 0.05):.2f}")
print(f"Commission at 8%: ${calculate_commission_good(1000, 0.08):.2f}")
print(f"Commission at 10%: ${calculate_commission_good(1000, 0.10):.2f}")

print("\n" + "="*60)
print("MISTAKE 3: Missing Return Paths")
print("="*60)

# ❌ BAD: Not all paths return a value
def validate_age_bad(age):
    if age >= 18:
        return "Adult"
    else:
        print("Not adult")  # Doesn't return!

# ✅ GOOD: All paths return a value
def validate_age_good(age):
    if age >= 18:
        return "Adult"
    else:
        return "Minor"

print("❌ BAD result:")
result1 = validate_age_bad(15)
print(f"validate_age_bad(15) returned: {result1}")  # Returns None
print(f"Can't use None in further calculations!")

print("\n✅ GOOD result:")
result2 = validate_age_good(15)
print(f"validate_age_good(15) returned: {result2}")  # Returns "Minor"
print(f"Can reliably use '{result2}' for further processing")

print("\n" + "="*60)
print("MISTAKE 4: Forgetting to Store Results")
print("="*60)

def calculate_total(price, quantity, tax_rate):
    subtotal = price * quantity
    tax = subtotal * tax_rate
    return subtotal + tax

print("❌ BAD (result is lost):")
calculate_total(50, 3, 0.1)
print("The result is computed but not stored - what a waste!")

print("\n✅ GOOD (result is stored for reuse):")
total = calculate_total(50, 3, 0.1)
print(f"Total: ${total:.2f}")
discount = total * 0.05
print(f"Student discount (5%): ${discount:.2f}")
final = total - discount
print(f"Final price: ${final:.2f}")
````

### Mistake 2: Hardcoded Values

Avoid hardcoding values inside functions. Use parameters instead.

````python
# Section 6: Passing Returned Values to Other Functions

# Data processing functions
def clean_text(text):
    """Clean text by removing extra spaces and converting to lowercase"""
    return text.strip().lower()

def count_words(text):
    """Count words in text"""
    words = text.split()
    return len(words)

def get_first_word(text):
    """Get the first word of text"""
    words = text.split()
    return words[0] if words else ""

def calculate_word_statistics(text):
    """Get multiple statistics from text"""
    word_count = count_words(text)
    first_word = get_first_word(text)
    char_count = len(text)
    return word_count, first_word, char_count

print("PASSING RETURNED VALUES AS ARGUMENTS:")
print("="*60)

raw_text = "  Hello World Python Programming  "
print(f"Original text: '{raw_text}'")

# Chain of functions
cleaned = clean_text(raw_text)
print(f"After clean_text(): '{cleaned}'")

num_words = count_words(cleaned)
print(f"After count_words(): {num_words} words")

first = get_first_word(cleaned)
print(f"After get_first_word(): '{first}'")

print("\n" + "="*60)
print("COMPLEX DATA FLOW:")
print("="*60)

# More complex example
def analyze_survey_responses(scores):
    """Analyze survey scores and return status"""
    avg = sum(scores) / len(scores)
    
    if avg >= 4.0:
        return "Excellent", avg
    elif avg >= 3.0:
        return "Good", avg
    else:
        return "Needs Improvement", avg

def format_survey_report(status, average_score):
    """Format the survey results into a report"""
    return f"Survey Status: {status} (Average: {average_score:.2f}/5.0)"

# Data flows through multiple functions
responses = [4.5, 3.8, 4.2, 4.1, 3.9]
print(f"Survey responses: {responses}")

# Returned values become arguments to next function
status, avg = analyze_survey_responses(responses)
report = format_survey_report(status, avg)
print(report)

print("\n" + "="*60)
print("DIRECT NESTING OF FUNCTION CALLS:")
print("="*60)

# You can also nest them directly
report = format_survey_report(*analyze_survey_responses(responses))
print(f"Nested approach: {report}")
````

## Section 6: Passing Returned Values to Other Functions

### Building Function Chains

Returned values from one function can become arguments to another function. This is how we build powerful, composable programs.

### Data flows: Input → Function 1 → Output → Input → Function 2 → Output

````python
# Section 5: Using Returned Results in Calculations

# Helper functions
def calculate_base_salary(hourly_rate, hours_per_week, weeks_per_year=52):
    """Calculate annual base salary"""
    return hourly_rate * hours_per_week * weeks_per_year

def apply_bonus(salary, bonus_percent):
    """Add a bonus to salary"""
    bonus = salary * (bonus_percent / 100)
    return salary + bonus

def calculate_tax(gross_income, tax_rate):
    """Calculate tax owed"""
    return gross_income * (tax_rate / 100)

print("USING RETURNED RESULTS IN CALCULATIONS:")
print("="*60)

# Step 1: Calculate base salary
hourly_rate = 40
hours_per_week = 40
weeks_per_year = 52

base_salary = calculate_base_salary(hourly_rate, hours_per_week, weeks_per_year)
print(f"Hourly Rate: ${hourly_rate}/hr")
print(f"Hours/Week: {hours_per_week}")
print(f"Weeks/Year: {weeks_per_year}")
print(f"Step 1 - Base Salary: ${base_salary:,.2f}")

# Step 2: Add bonus (use result from step 1)
bonus_percent = 15
salary_with_bonus = apply_bonus(base_salary, bonus_percent)
print(f"\nStep 2 - Apply {bonus_percent}% bonus: ${salary_with_bonus:,.2f}")

# Step 3: Calculate taxes (use result from step 2)
tax_rate = 22
taxes_owed = calculate_tax(salary_with_bonus, tax_rate)
print(f"\nStep 3 - Calculate taxes ({tax_rate}%): ${taxes_owed:,.2f}")

# Step 4: Final take-home pay
take_home = salary_with_bonus - taxes_owed
print(f"\nStep 4 - Take-home pay: ${take_home:,.2f}")

print("\n" + "="*60)
print("CHAINING WITH NESTED CALLS:")
print("="*60)

# You can also nest function calls
def gross_after_bonus_and_tax(hourly_rate, hours_per_week, bonus_pct, tax_rate):
    """End-to-end calculation in one call"""
    base = calculate_base_salary(hourly_rate, hours_per_week)
    with_bonus = apply_bonus(base, bonus_pct)
    taxes = calculate_tax(with_bonus, tax_rate)
    net = with_bonus - taxes
    return net

net_pay = gross_after_bonus_and_tax(35, 40, 10, 20)
print(f"\nCompact calculation in one call: ${net_pay:,.2f}")
````

## Section 5: Using Returned Results in Calculations

### Store Results in Variables

Always store returned values so you can use them later.

### Chain Results Through Calculations

Use returned values in further computations to build complex logic from simple functions.

````python
# Section 4: Returning Values from Functions

# Return a single value
def calculate_average(scores):
    """Calculate and return the average"""
    if len(scores) == 0:
        return 0
    return sum(scores) / len(scores)

print("RETURNING SINGLE VALUES:")
test_scores = [85, 90, 78, 92, 88]
avg = calculate_average(test_scores)
print(f"Test scores: {test_scores}")
print(f"Average returned: {avg:.2f}")
print()

# Return different types
def convert_temperature(celsius, to_fahrenheit=True):
    """Convert temperature and return the result"""
    if to_fahrenheit:
        return (celsius * 9/5) + 32
    else:
        return celsius * 1.8  # Just showing it can change

print("RETURNING DIFFERENT TYPES:")
temp_c = 25
temp_f = convert_temperature(temp_c)
print(f"{temp_c}°C = {temp_f:.1f}°F")
print()

# Return multiple values (tuple)
def get_student_stats(scores):
    """Return multiple statistics"""
    average = sum(scores) / len(scores)
    highest = max(scores)
    lowest = min(scores)
    return average, highest, lowest  # Returns a tuple

print("RETURNING MULTIPLE VALUES:")
grades = [92, 85, 78, 95, 88]
avg, high, low = get_student_stats(grades)
print(f"Scores: {grades}")
print(f"Average: {avg:.1f}, Highest: {high}, Lowest: {low}")
print()

# Return ends execution
def check_credentials(username, password):
    """Validate username and password"""
    if not username:
        return "Error: Username required"
    
    if len(password) < 8:
        return "Error: Password must be 8+ characters"  # Returns here
    
    # Code below never runs if either return above is triggered
    return "Login successful"

print("RETURN ENDS EXECUTION:")
print(check_credentials("alice", "short"))
print(check_credentials("", "longpassword123"))
print(check_credentials("bob", "strongpassword123"))
````
