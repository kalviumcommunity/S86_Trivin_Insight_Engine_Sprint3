# Milestone: Visualizing Data Distributions Using Boxplots

## Objective
Learn to create and interpret boxplots for visualizing data distributions, identifying outliers, and comparing spread across columns.

## What You'll Learn
- Understand what a boxplot represents
- Visualize distribution spread using quartiles
- Identify median and interquartile range (IQR)
- Detect potential outliers visually
- Compare distributions across multiple columns

## Key Concepts
**Boxplot Components:**
- **Median (Q2)**: The 50th percentile—the middle value
- **Lower Quartile (Q1)**: The 25th percentile—25% of data below this
- **Upper Quartile (Q3)**: The 75th percentile—75% of data below this
- **IQR (Interquartile Range)**: Q3 - Q1, representing the middle 50% of data
- **Whiskers**: Lines extending from the box (typically 1.5 × IQR from quartiles)
- **Outliers**: Points beyond the whiskers, shown as individual dots

## Why Boxplots Matter
- **Compact Summary**: Summarize distributions in a single, easy-to-read visualization
- **Outlier Detection**: Visually identify potential outliers without calculating them manually
- **Comparison**: Compare distributions across multiple columns or groups side-by-side
- **Complementary to Histograms**: Boxplots show quartiles and outliers clearly, while histograms show frequency distribution
- **Better EDA**: Make informed data cleaning and analysis decisions

## Setup: Import Libraries and Load Data

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the employee survey data
file_path = '../data/raw/employee_survey_2026_Q1.csv'
df = pd.read_csv(file_path)

# Display basic information
print("Dataset shape:", df.shape)
print("\nFirst few rows:")
print(df.head())
print("\nColumn names and types:")
print(df.dtypes)
print("\nBasic statistics:")
print(df.describe())
````

## Section 1: Understanding Boxplots

A boxplot is a standardized way of displaying the distribution of data based on quartiles. Let's understand each component:

### Understanding Boxplot Components

Let's examine a single column and calculate its quartiles manually first:

````python
# Select a numeric column
column = 'satisfaction_score'
data = df[column]

# Calculate key statistics
q1 = data.quantile(0.25)  # 25th percentile
median = data.quantile(0.5)  # 50th percentile (median)
q3 = data.quantile(0.75)  # 75th percentile
iqr = q3 - q1  # Interquartile range

# Whisker boundaries
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr

print(f"Column: {column}")
print(f"\nQuartile Analysis:")
print(f"  Minimum value: {data.min():.2f}")
print(f"  Q1 (25th percentile): {q1:.2f}")
print(f"  Median (Q2, 50th percentile): {median:.2f}")
print(f"  Q3 (75th percentile): {q3:.2f}")
print(f"  Maximum value: {data.max():.2f}")
print(f"\nSpread Analysis:")
print(f"  IQR (Q3 - Q1): {iqr:.2f}")
print(f"  Lower whisker boundary: {lower_whisker:.2f}")
print(f"  Upper whisker boundary: {upper_whisker:.2f}")

# Identify outliers
outliers = data[(data < lower_whisker) | (data > upper_whisker)]
print(f"\nOutliers (beyond whiskers): {len(outliers)} values")
if len(outliers) > 0:
    print(f"  Outlier values: {sorted(outliers.values)}")
````

## Section 2: Creating a Boxplot for a Single Column

Let's create a simple boxplot to visualize one distribution.

````python
# Create a figure with a single boxplot
fig, ax = plt.subplots(figsize=(8, 6))

# Create boxplot for satisfaction_score
ax.boxplot(df['satisfaction_score'], vert=True, patch_artist=True,
           boxprops=dict(facecolor='lightblue', color='darkblue'),
           whiskerprops=dict(color='darkblue'),
           capprops=dict(color='darkblue'),
           medianprops=dict(color='red', linewidth=2),
           flierprops=dict(marker='o', markerfacecolor='red', markersize=8, alpha=0.6))

ax.set_ylabel('Satisfaction Score', fontsize=12)
ax.set_title('Distribution of Satisfaction Scores', fontsize=14, fontweight='bold')
ax.set_xticklabels(['Satisfaction Score'])
ax.grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.savefig('../outputs/figures/01_single_boxplot_satisfaction.png', dpi=100, bbox_inches='tight')
plt.show()

print("Boxplot created successfully!")
````

### Interpreting the Single Column Boxplot

Look at the boxplot above and answer these questions:
1. **What is the median satisfaction score?** (Look for the red line in the middle of the box)
2. **Is the distribution symmetric?** (Is the red line in the center of the box, or off to one side?)
3. **Are there any outliers?** (Red dots beyond the whiskers)
4. **What does the box height tell us?** (Height = IQR, showing where middle 50% of data lies)

````python
# Let's verify our observations with statistics
satisfaction_data = df['satisfaction_score']

print(f"Satisfaction Score Analysis:")
print(f"  Mean: {satisfaction_data.mean():.2f}")
print(f"  Median: {satisfaction_data.median():.2f}")
print(f"  Std Dev: {satisfaction_data.std():.2f}")
print(f"  Range: {satisfaction_data.min():.0f} to {satisfaction_data.max():.0f}")
print(f"  IQR: {satisfaction_data.quantile(0.75) - satisfaction_data.quantile(0.25):.2f}")

# Check for skewness (median vs mean)
if satisfaction_data.median() > satisfaction_data.mean():
    print(f"  Distribution: Left-skewed (median > mean)")
elif satisfaction_data.median() < satisfaction_data.mean():
    print(f"  Distribution: Right-skewed (median < mean)")
else:
    print(f"  Distribution: Symmetric (median ≈ mean)")
````

## Section 3: Comparing Boxplots Across Multiple Columns

One of the most powerful uses of boxplots is comparing distributions side-by-side. Let's compare multiple survey metrics.

````python
# Select all numeric columns (survey metrics)
numeric_columns = ['satisfaction_score', 'work_life_balance', 'managment_support', 
                   'career_growth', 'team_collaboration']

# Check if these columns exist
available_columns = [col for col in numeric_columns if col in df.columns]
print(f"Comparing {len(available_columns)} metrics: {available_columns}")

# Create a figure with multiple boxplots
fig, ax = plt.subplots(figsize=(12, 6))

# Create boxplots for all metrics
bp = ax.boxplot([df[col] for col in available_columns], 
                 labels=available_columns,
                 vert=True,
                 patch_artist=True,
                 widths=0.6)

# Customize colors
colors = ['lightblue', 'lightgreen', 'lightyellow', 'lightcoral', 'lightpink']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_edgecolor('darkblue')

# Customize median lines
for median in bp['medians']:
    median.set_color('red')
    median.set_linewidth(2)

# Customize outliers
for flier in bp['fliers']:
    flier.set(marker='o', markerfacecolor='red', markersize=6, alpha=0.6)

ax.set_ylabel('Score', fontsize=12)
ax.set_title('Comparison of Employee Satisfaction Metrics', fontsize=14, fontweight='bold')
ax.set_ylim(0, 11)
plt.xticks(rotation=45, ha='right')
ax.grid(axis='y', alpha=0.3, linestyle='--')

plt.tight_layout()
plt.savefig('../outputs/figures/02_comparison_boxplots_all_metrics.png', dpi=100, bbox_inches='tight')
plt.show()

print("\nComparison boxplot created!")
````

### Analyzing the Comparison

Now that we have all metrics side-by-side, let's compare them:

````python
# Create a comparison table
comparison_stats = pd.DataFrame()

for col in available_columns:
    data = df[col]
    comparison_stats[col] = {
        'Mean': data.mean(),
        'Median': data.median(),
        'Q1': data.quantile(0.25),
        'Q3': data.quantile(0.75),
        'IQR': data.quantile(0.75) - data.quantile(0.25),
        'Min': data.min(),
        'Max': data.max(),
        'Std Dev': data.std()
    }

comparison_stats = comparison_stats.T.round(2)
print("Statistical Comparison of All Metrics:")
print(comparison_stats)

# Identify key insights
print("\n" + "="*60)
print("KEY INSIGHTS FROM BOXPLOT COMPARISON")
print("="*60)

# Find highest and lowest medians
highest_median_col = comparison_stats['Median'].idxmax()
lowest_median_col = comparison_stats['Median'].idxmin()
print(f"\n1. Central Tendency:")
print(f"   - Highest median: {highest_median_col} ({comparison_stats.loc[highest_median_col, 'Median']:.2f})")
print(f"   - Lowest median: {lowest_median_col} ({comparison_stats.loc[lowest_median_col, 'Median']:.2f})")

# Find widest and narrowest spreads (IQR)
widest_iqr_col = comparison_stats['IQR'].idxmax()
narrowest_iqr_col = comparison_stats['IQR'].idxmin()
print(f"\n2. Spread (IQR):")
print(f"   - Most variable: {widest_iqr_col} (IQR: {comparison_stats.loc[widest_iqr_col, 'IQR']:.2f})")
print(f"   - Most consistent: {narrowest_iqr_col} (IQR: {comparison_stats.loc[narrowest_iqr_col, 'IQR']:.2f})")

# Find columns with the widest overall range
comparison_stats['Range'] = comparison_stats['Max'] - comparison_stats['Min']
widest_range_col = comparison_stats['Range'].idxmax()
print(f"\n3. Overall Range:")
print(f"   - Widest range: {widest_range_col} ({comparison_stats.loc[widest_range_col, 'Min']:.0f}-{comparison_stats.loc[widest_range_col, 'Max']:.0f})")
````

## Section 5: Interpreting Outliers Carefully

Outliers are points beyond 1.5 × IQR from the quartiles. But remember: **outliers are not always errors!**

### What Outliers Tell Us (And What They Don't)

**Outliers CAN indicate:**
- Unusual or extreme values that deserve investigation
- Data entry errors or measurement problems
- Exceptional cases that buck the trend
- Groups within your data that behave differently

**Outliers SHOULD NOT be:**
- Automatically removed without investigation
- Assumed to be errors just because they're extreme
- Ignored if they reveal important insights
- Used to make decisions without context

````python
# Function to identify outliers using IQR method
def find_outliers(data, column_name):
    """
    Identify outliers using the IQR method (values beyond 1.5 × IQR from quartiles).
    """
    Q1 = data[column_name].quantile(0.25)
    Q3 = data[column_name].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outlier_indices = (data[column_name] < lower_bound) | (data[column_name] > upper_bound)
    outliers = data[outlier_indices]
    
    return outliers, lower_bound, upper_bound

# Find outliers in satisfaction_score
outliers, lower, upper = find_outliers(df, 'satisfaction_score')

print(f"Satisfaction Score Outlier Analysis")
print(f"="*60)
print(f"Lower bound (Q1 - 1.5×IQR): {lower:.2f}")
print(f"Upper bound (Q3 + 1.5×IQR): {upper:.2f}")
print(f"\nNumber of outliers: {len(outliers)} out of {len(df)} records ({len(outliers)/len(df)*100:.1f}%)")

if len(outliers) > 0:
    print(f"\nOutlier records:")
    print(outliers[['employee_id', 'department', 'satisfaction_score', 'response_text']].to_string())
else:
    print("\nNo outliers found in satisfaction_score!")
````

````python
# Analyze outliers across all metrics
print("\nOutlier Summary Across All Metrics:")
print("="*60)

metrics_to_check = available_columns
outlier_summary = {}

for metric in metrics_to_check:
    outliers, lower, upper = find_outliers(df, metric)
    outlier_summary[metric] = {
        'count': len(outliers),
        'percentage': f"{len(outliers)/len(df)*100:.1f}%",
        'lower_bound': f"{lower:.2f}",
        'upper_bound': f"{upper:.2f}"
    }

outlier_df = pd.DataFrame(outlier_summary).T
print(outlier_df.to_string())

# Identify metric with most outliers
metric_with_most_outliers = max(outlier_summary.items(), key=lambda x: x[1]['count'])[0]
print(f"\nMetric with most outliers: {metric_with_most_outliers}")
````

### How to Handle Outliers: A Decision Framework

| Scenario | Action | Reason |
|----------|--------|--------|
| Data entry error (e.g., 99 instead of 9) | Correct or remove | Clear mistake, fix the problem |
| Legitimate extreme value | Keep & investigate | May reveal important insights |
| Different subgroup (e.g., department-level difference) | Keep & analyze by group | Outlier may be normal within subgroup |
| Measurement error | Remove with documentation | Unreliable data, mark in analysis notes |
| You don't understand why it's there | INVESTIGATE FIRST | Never remove blindly |

**Golden Rule: Document your decisions and the reasoning behind them.**

````python
# Investigate a specific outlier
print("\nInvestigating Outliers - Example Analysis")
print("="*60)

# Find employees with extreme satisfaction scores
very_satisfied = df[df['satisfaction_score'] >= 9]
very_dissatisfied = df[df['satisfaction_score'] <= 3]

print(f"\nVery Satisfied Employees (Score ≥ 9): {len(very_satisfied)}")
print(very_satisfied[['employee_id', 'department', 'satisfaction_score', 'response_text']].head(3).to_string())

print(f"\nVery Dissatisfied Employees (Score ≤ 3): {len(very_dissatisfied)}")
print(very_dissatisfied[['employee_id', 'department', 'satisfaction_score', 'response_text']].head(3).to_string())

print(f"\nQ: Are these outliers or legitimate extreme cases?")
print(f"A: They appear to be legitimate survey responses reflecting genuine sentiment.")
print(f"   They should be KEPT for analysis, but investigated for patterns.")

# Department pattern analysis
print(f"\nDepartment distribution of extreme scores:")
print(f"\nVery Satisfied by Department:")
print(very_satisfied['department'].value_counts())
print(f"\nVery Dissatisfied by Department:")
print(very_dissatisfied['department'].value_counts())
````

## Section 7: Key Takeaways

### What You've Learned

1. **Boxplot Components**: Median, quartiles, IQR, whiskers, and outliers each tell part of the distribution story

2. **Single vs. Multiple**: Boxplots are powerful for comparing multiple distributions side-by-side

3. **Grouping**: Boxplots work great for comparing groups (departments, categories, etc.)

4. **Outlier Context**: Always investigate outliers in context before removing them

5. **Complementary Tool**: Use boxplots alongside histograms, statistics, and domain knowledge

### Common Beginner Mistakes to Avoid

❌ **Removing outliers without investigation** - They might be valuable insights

❌ **Using only averages** - Boxplots reveal spread and distribution shape that means can hide

❌ **Ignoring scale differences** - Be careful when comparing metrics on different scales

❌ **Not documenting decisions** - Always record why you kept or removed data

❌ **Assuming outliers = errors** - Context matters; sometimes outliers are the most interesting data

### When to Use Boxplots in Your Analysis

✅ Early EDA (exploratory data analysis)

✅ Comparing distributions across groups or categories

✅ Identifying data quality issues and outliers

✅ Communicating distributions to non-technical stakeholders (clear visuals)

✅ Assessing if data meets assumptions for statistical tests
