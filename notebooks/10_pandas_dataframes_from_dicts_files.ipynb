## Section 1: Understanding Pandas DataFrames

### What is a DataFrame?
A **DataFrame** is a two-dimensional data structure in Pandas that represents tabular data similar to:
- Excel spreadsheets
- Database tables
- CSV files

### Key Components
- **Rows**: Individual records or observations (often labeled by an index)
- **Columns**: Features or attributes (each with a name and consistent data type)
- **Index**: Row labels (0, 1, 2... by default, but can be custom)
- **Data Types**: Each column has a specific type (int, float, string, datetime, etc.)

### DataFrame Structure Example
```
        Name          Age       City       Salary
Index
0      Alice          28       NYC        75000
1      Bob            35       LA         85000
2      Carol          42       Chicago    95000
```

### DataFrame vs Series
- **Series**: 1-dimensional (single column) with index and values
- **DataFrame**: 2-dimensional (multiple columns) with index, column names, and values

````python
# Example: Creating a simple DataFrame manually to understand structure
employees_data = {
    'Name': ['Alice', 'Bob', 'Carol'],
    'Age': [28, 35, 42],
    'City': ['NYC', 'LA', 'Chicago'],
    'Salary': [75000, 85000, 95000]
}

df_example = pd.DataFrame(employees_data)
print("Sample DataFrame:")
print(df_example)
print("\n" + "="*60)
print("DataFrame shape (rows, columns):", df_example.shape)
print("Column names:", df_example.columns.tolist())
print("Index:", df_example.index.tolist())
print("Data types:\n", df_example.dtypes)
````

## Section 2: Creating DataFrames from Dictionaries

### Dictionary-Based DataFrame Creation
Dictionaries are a natural way to create DataFrames programmatically. Each key becomes a column name, and each value becomes the column data.

### Format: Column-Style Dictionary
```python
data = {
    'column_name1': [val1, val2, val3],
    'column_name2': [val1, val2, val3],
    ...
}
df = pd.DataFrame(data)
```

**Key Notes:**
- All lists must have the same length
- Dictionary keys become column names
- List values become column data (rows)
- Pandas automatically creates a default integer index (0, 1, 2...)

````python
# Example 1: Create a DataFrame from a column-style dictionary
student_data = {
    'Student': ['Alice', 'Bob', 'Carol', 'David'],
    'Math': [95, 87, 92, 88],
    'Science': [89, 91, 94, 90],
    'History': [85, 88, 96, 92]
}

df_students = pd.DataFrame(student_data)
print("Example 1: Student Grades DataFrame")
print(df_students)
print("\nDataFrame Info:")
print(f"  Shape: {df_students.shape}")
print(f"  Columns: {df_students.columns.tolist()}")
print(f"  Data types:\n{df_students.dtypes}")
````

````python
# Example 2: Accessing columns and rows from a dictionary-created DataFrame
print("\n" + "="*60)
print("Example 2: Accessing Data from Dictionary DataFrame")
print("="*60)

# Access a single column
print("\nAccessing 'Math' column:")
print(df_students['Math'])

# Access a specific cell
print("\nValue at row 2, column 'Science':")
print(df_students.loc[2, 'Science'])

# Access first row
print("\nFirst row:")
print(df_students.iloc[0])

# Access multiple columns
print("\nMultiple columns (Student and Math):")
print(df_students[['Student', 'Math']])
````

## Section 3: Creating DataFrames from Files

### File-Based Data Loading
Loading data from external files is the most common real-world scenario. Pandas supports multiple file formats:
- **CSV** (Comma-Separated Values) - most common
- **Excel** (.xlsx, .xls)
- **JSON** (JavaScript Object Notation)
- **SQL** databases
- **parquet**, **HDF5**, and others

### Loading CSV Files
```python
df = pd.read_csv('filename.csv')
```

**Common Parameters:**
- `sep` or `delimiter`: Column separator (default: ',')
- `header`: Row number for column names (default: 0)
- `index_col`: Column to use as index
- `dtype`: Specify data types for columns
- `nrows`: Load only first n rows (useful for large files)

### Why File Loading Matters
- Real-world data is almost always in external files
- Correct file loading prevents cascading errors
- Understanding headers and encoding is essential
- File size considerations affect memory usage

````python
# Example 3: Load a DataFrame from a CSV file
print("\n" + "="*60)
print("Example 3: Loading Data from CSV File")
print("="*60)

# Path to the CSV file
csv_file = '../data/raw/employee_survey_2026_Q1.csv'

# Load the CSV into a DataFrame
df_employees = pd.read_csv(csv_file)

print(f"\nDataFrame loaded from: {csv_file}")
print(f"File exists: {os.path.exists(csv_file)}")
print(f"\nFirst few rows of the employee survey data:")
print(df_employees.head())

print(f"\nShape of loaded DataFrame: {df_employees.shape}")
print(f"Number of rows: {df_employees.shape[0]}")
print(f"Number of columns: {df_employees.shape[1]}")
````

````python
# Example 4: Loading CSV with specific parameters
print("\n" + "="*60)
print("Example 4: Loading CSV with Parameters")
print("="*60)

# Load only first 5 rows
df_sample = pd.read_csv(csv_file, nrows=5)
print(f"\nLoading only first 5 rows:")
print(f"Shape: {df_sample.shape}")
print(df_sample)

# Load and check column names
df_check = pd.read_csv(csv_file)
print(f"\nColumn names in the file:")
for i, col in enumerate(df_check.columns):
    print(f"  {i}: {col}")
    
print(f"\nData types of columns:")
print(df_check.dtypes)
````

## Section 4: Inspecting DataFrame Structure

### Why Inspect DataFrames?
After creating or loading a DataFrame, always inspect it to:
- Verify data loaded correctly
- Understand shape and dimensions
- Check data types
- Identify missing values
- Prevent silent errors in analysis

### Essential Inspection Methods

| Method | Purpose |
|--------|---------|
| `.head()` | View first rows |
| `.tail()` | View last rows |
| `.shape` | Get (rows, columns) |
| `.columns` | Get column names |
| `.dtypes` | Get data types |
| `.info()` | Get detailed structure |
| `.describe()` | Get statistical summary |
| `.isnull()` | Identify missing values |
| `.index` | Get row index labels |

### Best Practice
**Always inspect your data immediately after loading!** This catches errors early and builds confidence in your data pipeline.

````python
# Example 5: Inspecting DataFrame structure using employee data
print("\n" + "="*60)
print("Example 5: Inspecting DataFrame Structure")
print("="*60)

df = pd.read_csv(csv_file)

# 1. View first rows (default 5 rows)
print("\n1. First 5 rows (.head()):")
print(df.head())

# 2. View last rows
print("\n2. Last 3 rows (.tail(3)):")
print(df.tail(3))

# 3. DataFrame shape
print(f"\n3. DataFrame shape (.shape): {df.shape}")
print(f"   Total rows: {df.shape[0]}")
print(f"   Total columns: {df.shape[1]}")

# 4. Column names
print(f"\n4. Column names (.columns):")
print(df.columns.tolist())
````

````python
# Example 6: More detailed inspection methods
print("\n" + "="*60)
print("Example 6: Detailed Inspection Methods")
print("="*60)

# 5. Data types
print("\n5. Data types (.dtypes):")
print(df.dtypes)

# 6. DataFrame info (comprehensive overview)
print("\n6. Complete DataFrame info (.info()):")
df.info()

# 7. Statistical summary
print("\n7. Statistical summary (.describe()):")
print(df.describe())
````

````python
# Example 7: Checking for missing values
print("\n" + "="*60)
print("Example 7: Detecting Missing Values")
print("="*60)

# Check for missing values
print("\nMissing values per column (.isnull().sum()):")
print(df.isnull().sum())

# Total missing values
print(f"\nTotal missing values in DataFrame: {df.isnull().sum().sum()}")

# Percentage of missing values
print("\nPercentage of missing values per column:")
missing_pct = (df.isnull().sum() / len(df)) * 100
print(missing_pct[missing_pct > 0])
````

## Common Issues and Solutions

### Issue 1: Length Mismatch in Dictionary
```python
# ❌ ERROR: Lists have different lengths
data = {'A': [1, 2, 3], 'B': [1, 2]}  # B is too short!
df = pd.DataFrame(data)  # ValueError!

# ✓ CORRECT: All lists same length
data = {'A': [1, 2, 3], 'B': [1, 2, 3]}
df = pd.DataFrame(data)
```

### Issue 2: File Not Found
```python
# ❌ ERROR: Wrong path
df = pd.read_csv('data.csv')  # FileNotFoundError if file doesn't exist

# ✓ CORRECT: Verify path exists
import os
file_path = 'data/raw/data.csv'
if os.path.exists(file_path):
    df = pd.read_csv(file_path)
else:
    print(f"File not found: {file_path}")
```

### Issue 3: Confusing Series and DataFrame
```python
# Series: 1D (single column)
series = df['ColumnName']  # Type: Series
print(type(series))  # <class 'pandas.core.series.Series'>

# DataFrame: 2D (multiple columns or single column with brackets)
df_single = df[['ColumnName']]  # Type: DataFrame (notice double brackets)
print(type(df_single))  # <class 'pandas.core.frame.DataFrame'>
```

### Best Practice Checklist
- [ ] Always inspect data immediately after loading
- [ ] Verify file exists before attempting to load
- [ ] Check column names match expectations
- [ ] Verify data types are correct
- [ ] Check for unexpected missing values
- [ ] Use `nrows` parameter for large files during development
- [ ] Keep file paths relative when possible
