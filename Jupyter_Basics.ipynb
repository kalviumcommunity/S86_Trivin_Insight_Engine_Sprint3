# Running, Restarting, and Interrupting Jupyter Kernels

This section focuses on understanding how Jupyter kernels work and how to control them safely.

**What is a Jupyter Kernel?**

A kernel is the **computational engine** that runs your code. Think of it as the "brain" of your notebook:
- It stores all variables in memory
- It remembers the execution history
- It keeps track of imported libraries
- It maintains state between cell executions

**Why Kernel Management Matters:**

Many notebook issues are NOT caused by code errors, but by **kernel state problems**:
- ❌ Cells run out of order, creating dependency issues
- ❌ Stale variables with unexpected values
- ❌ Long-running executions that freeze the notebook
- ❌ Hidden state that makes results non-reproducible

**What You'll Learn:**
1. How to run cells deliberately and observe execution order
2. How to restart the kernel to reset state
3. How to interrupt long-running or stuck cells
4. When to use restart vs interrupt

## Part 1: Running Cells and Understanding Execution Order

The kernel executes cells in the **order you run them**, not the order they appear in the notebook.

This can create hidden dependencies and unexpected behavior.

### Example: Order Matters

```python
# Cell 1: Define a variable
x = 10
print(f"x is set to: {x}")
```

```python
# Cell 3: Use the variable
result = x + 5
print(f"x + 5 = {result}")
```

### Real-World Example: Function Redefinition

This scenario happens often during development:

```python
# Step B: Accidentally redefine with a bug
def calculate_discount(price, discount_rate):
    return price * discount_rate  # Bug: forgot the (1 - ...)

# If you run this cell, the function is now broken!
result = calculate_discount(100, 0.2)
print(f"Price after discount: ${result}")  # Wrong! Should be 80, not 20
```

## Part 2: Restarting the Kernel

Restarting the kernel **clears all variables and memory**. It's like closing and reopening your notebook with a fresh start.

**When to restart:**
- Variables have unexpected values
- You want to test if your notebook runs from top to bottom
- You've run cells out of order and need a clean slate
- Before sharing your notebook with others

**How to restart in VS Code:**
- Click the **↻ Restart** button in the notebook toolbar
- Or use the command palette: `Jupyter: Restart Kernel`
- After restarting, all variables are cleared

```python
# Example: Long-running operation
# Run this cell, then interrupt it after a few seconds

import time

print("Starting long operation...")
print("(Interrupt this by clicking the Stop button ■)")
print()

for i in range(100):
    print(f"Processing item {i+1}/100...", end='\r')
    time.sleep(0.5)  # Simulates slow operation
    
print("\nOperation completed!")
```

```python
# Verify that interrupting doesn't clear variables
try:
    print(f"x is still defined: {x}")
    print("✅ Interrupt preserved kernel state!")
except NameError:
    print("❌ Variable was cleared (this shouldn't happen with interrupt)")
```

## Part 4: Restart vs Interrupt - When to Use Each

Understanding the difference is crucial for efficient debugging and development.

| Action | What It Does | When to Use | Variables Preserved? |
|--------|-------------|-------------|---------------------|
| **Interrupt** | Stops current cell execution | Cell taking too long or stuck | ✅ Yes |
| **Restart** | Clears all memory and variables | Need fresh start or test reproducibility | ❌ No |

### Decision Tree:

**Is a cell currently running and you want to stop it?**
- → Use **Interrupt** (■ Stop button)

**Are your variables in a messy state or you want to test from scratch?**
- → Use **Restart** (↻ Restart button)

**Did you run cells out of order and get unexpected results?**
- → Use **Restart**, then run cells sequentially from top

**Want to verify your notebook works for others?**
- → Use **Restart**, then "Run All Cells"

### Common Scenarios

**Scenario 1: "My code worked before, but now it's broken"**
- **Likely cause:** Cells run out of order, variables have unexpected values
- **Solution:** Restart kernel, run all cells from top

**Scenario 2: "This cell is taking forever"**
- **Likely cause:** Infinite loop, slow operation, or large computation
- **Solution:** Interrupt execution, fix the code, try again

**Scenario 3: "I want to make sure my notebook is reproducible"**
- **Best practice:** Restart kernel → "Run All Cells"
- If all cells run successfully top-to-bottom, your notebook is reproducible!

**Scenario 4: "I'm getting weird errors about undefined variables"**
- **Likely cause:** Kernel state doesn't match what you expect
- **Solution:** Restart kernel, run cells in correct order from the top

## Part 5: Best Practices for Kernel Management

Follow these practices to avoid common notebook problems:

### ✅ **DO:**
- ✅ Run cells in sequential order from top to bottom
- ✅ Restart and run all cells before sharing your notebook
- ✅ Test reproducibility by restarting frequently
- ✅ Use interrupt for stuck or long-running cells
- ✅ Check execution numbers to spot out-of-order execution
- ✅ Document dependencies between cells clearly

### ❌ **DON'T:**
- ❌ Run cells randomly out of order
- ❌ Rely on variables defined in cells you haven't run
- ❌ Let cells run indefinitely without interrupting
- ❌ Share notebooks without testing with a fresh kernel
- ❌ Ignore execution numbers (they reveal hidden problems)

### 🎯 **Pro Tip: Check Execution Numbers**

Look at the `[number]` next to each code cell:
- `[ ]` = Cell hasn't been run yet
- `[*]` = Cell is currently running
- `[5]` = Cell was the 5th to execute

**If these numbers are NOT sequential (1, 2, 3, 4...), you've run cells out of order!**

This is a warning sign—consider restarting and running from the top.

## Part 3: Interrupting Execution

Sometimes a cell takes too long or gets stuck. Interrupting stops execution **without losing your kernel state**.

**When to interrupt:**
- A cell is taking longer than expected
- You made a mistake (like an infinite loop)
- You want to stop and modify code
- You accidentally started a slow operation

**How to interrupt in VS Code:**
- Click the **■ Stop** button next to the running cell
- Or use the command palette: `Jupyter: Interrupt`
- The kernel state is preserved—only the current cell stops

```python
# Test: Check if variables exist after restart
try:
    print(f"x = {x}")
    print(f"result = {result}")
    print("✅ Variables are still in memory")
except NameError as e:
    print("❌ Variables have been cleared!")
    print(f"Error: {e}")
```

### 🚨 **Problem:**

If you run Step A (correct function), then Step B (broken function), the kernel remembers the broken version!

All future calculations will be **wrong**—even if you don't run Step B again.

**Solution:** Restart the kernel to clear all definitions and start fresh.

```python
# Step A: Define a function correctly
def calculate_discount(price, discount_rate):
    return price * (1 - discount_rate)

# Test it
result = calculate_discount(100, 0.2)
print(f"Price after 20% discount: ${result}")
```
