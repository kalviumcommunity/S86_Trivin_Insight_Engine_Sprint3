# Understanding Jupyter Notebook Cells: Code vs Markdown

This notebook teaches you the fundamental concepts of working with Jupyter Notebooks by distinguishing between **Code cells** and **Markdown cells**.

**Learning Objectives:**
- Understand the purpose of Code cells and Markdown cells
- Learn when to use each cell type
- Practice creating professional, readable notebooks
- Structure notebooks for clarity and explanation

## Part 1: Understanding Code Cells

**Code Cells** contain Python code that gets executed when you run them. This is where your logic, calculations, and computations happen.

Key points about Code cells:
- They execute Python code when you press Run (or Shift+Enter)
- They show output below the cell
- They should contain logic and operations
- Comments inside code cells explain technical details, not the big picture

```python
# Example Code Cell 1: Simple calculations
numbers = [5, 10, 15, 20, 25]
total = sum(numbers)
average = total / len(numbers)

print(f"Numbers: {numbers}")
print(f"Sum: {total}")
print(f"Average: {average}")
```

### What Just Happened?

The Code cell above:
1. Created a list of numbers
2. Calculated the sum using Python's built-in `sum()` function
3. Calculated the average by dividing sum by the number of items
4. Printed the results to display the output

This is what a Code cell does - **it executes and produces output**. Notice that this explanation is in a **Markdown cell**, not buried in code comments. This separation makes the notebook more readable and professional.

## Part 2: Understanding Markdown Cells

**Markdown Cells** contain formatted text, explanations, and documentation. They don't execute code.

Key points about Markdown cells:
- They contain explanatory text, not executable code
- You can use headings, bullet points, bold, italic, and links
- They should explain the "why" behind your code
- They organize your notebook into logical sections
- They make your notebook readable to others

This cell you're reading right now is a Markdown cell! It's explaining concepts, not running any computations.

```python
# Example Code Cell 2: Working with strings
name = "Data Scientist"
greeting = f"Welcome to Jupyter, {name}!"

print(greeting)
print("=" * len(greeting))
print(f"This message was created using Python code in a Code cell.")
```

### Why Separate Code from Explanation?

Notice the pattern in this notebook:
- **Markdown cell**: Explains what we're about to do
- **Code cell**: Actually does it
- **Markdown cell**: Explains what happened and what it means

This structure has major advantages:
- **Readability**: Someone can understand your work without reading code comments
- **Professionalism**: Clear structure shows organized thinking
- **Reviewability**: Teammates can quickly understand your approach
- **Maintenance**: Future you will understand what past you was thinking

## Key Takeaways

### ✅ Code Cells
- Purpose: Execute Python code
- Output: Shows results of computations
- Use for: Logic, calculations, data operations
- Example: `x = 5 + 3` then `print(x)`

### ✅ Markdown Cells
- Purpose: Explain and document
- Output: Formatted text (no code execution)
- Use for: Headings, explanations, descriptions, reasoning
- Example: "This code calculates the average of a list"

### ✅ Professional Notebook Structure
1. Start with a title (Markdown)
2. Explain what you'll do (Markdown)
3. Show the code that does it (Code)
4. Explain the results (Markdown)
5. Repeat for each logical section

**Remember**: Code shows *what* you did. Markdown explains *why* you did it. Together, they create professional Data Science work that others can understand and build upon.
